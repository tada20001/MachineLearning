## ÎèÖÎ¶Ω ÌòºÌï© Î™®Ìòï(Independent Mixuture Model)

ÎèÖÎ¶Ω ÌòºÌï© Î™®Ìòï(Independent Mixuture Model)ÏùÄ Ïó∞ÏÜçÌôïÎ•† Î≥ÄÏàòÏù¥ÏßÄÎßå Îã®ÏùºÌïú ÌôïÎ•†Î∂ÑÌè¨Î•º Í∞ÄÏßÄÏßÄ ÏïäÍ≥† Î≥µÏàòÏùò Ïó∞ÏÜç ÌôïÎ•† Î∂ÑÌè¨ Ï§ë ÌïòÎÇòÎ•º ÌôïÎ•†Ï†ÅÏúºÎ°ú ÏÑ†ÌÉùÌïòÎäî Î™®ÌòïÏùÑ ÎßêÌïúÎã§. 

#### Ïù¥ Îïå Ïó∞ÏÜçÌôïÎ•† Î∂ÑÌè¨Ïùò ÏÑ†ÌÉùÏùÄ ÎèÖÎ¶ΩÏ†ÅÏù∏ Ïù¥ÏÇ∞ ÌôïÎ•†Î∂ÑÌè¨Î•º ÏÇ¨Ïö©ÌïúÎã§.

\begin{eqnarray*}
p(x) 
&=& \sum_{i=1}^m {P}(C=i)\cdot {P}(X=x|C=i) \\
&=& \sum_{i=1}^m \delta_i p_i(x)
\end{eqnarray*}

+ ùëù(ùë•)  : Ï†ÑÏ≤¥ Independent Mixuture Î∂ÑÌè¨
+ ùëù_ùëñ (ùë•) : Independent MixutureÏùò Í∞Å ÏÑ±Î∂Ñ(component)Ïù¥ ÎêòÎäî Í∞úÎ≥ÑÏ†ÅÏù∏ Ïó∞ÏÜç ÌôïÎ•†Î∂ÑÌè¨
+ ùõø_ùëñ : mixing parameter. ÌäπÏ†ïÏãúÍ∞ÑÏóê ÎåÄÌï¥ Î™®Îì† ÏÑ±Î∂Ñ Ï§ë ÌäπÏ†ïÌïú ùëù_ùëñ (ùë•)Í∞Ä ÏÑ†ÌÉùÎê† ÌôïÎ•†. Ïù¥ÏÇ∞ ÌôïÎ•† Î∂ÑÌè¨
+ ‚àë ùõø_ùëñ = 1 : mixing parameterÏóê ÎåÄÌïú ÌôïÎ•† Ï†úÌïú Ï°∞Í±¥

### ÎèÖÎ¶Ω ÌòºÌï© Î™®ÌòïÏùò Ïòà: Î≤†Î•¥ÎàÑÏù¥-Ï†ïÍ∑ú ÌòºÌï© Î™®Ìòï

#### Î≤†Î•¥ÎàÑÏù¥-Ï†ïÍ∑ú ÌòºÌï©Î™®Ìòï(Bernoulli Normal-Mixuture Model)ÏùÄ Î≤†Î•¥ÎàÑÏù¥ ÌôïÎ•† Î≥ÄÏàòÏùò Í∞íÏóê Îî∞Îùº Îëê Í∞úÏùò ÏÑúÎ°ú Îã§Î•∏ Ïó∞ÏÜç Ï†ïÍ∑ú Î∂ÑÌè¨ Ï§ë ÌïòÎÇòÎ•º ÏÑ†ÌÉùÌïòÎäî ÌôïÎ•† Î∂ÑÌè¨Ïù¥Îã§.

## ÎßàÏΩîÌîÑ Ï≤¥Ïù∏(Markov Chain)

#### ÎßàÏΩîÌîÑ Ï≤¥Ïù∏ÏùÄ Îã§ÏùåÍ≥º Í∞ôÏùÄ ÎßàÏΩîÌîÑ(Markov) ÌäπÏÑ±ÏùÑ Í∞ÄÏßÄÎäî Ïù¥ÏÇ∞ÏãúÍ∞Ñ ÌôïÎ•† ÌîÑÎ°úÏÑ∏Ïä§Î•º ÎßêÌïúÎã§.

$$P(C_{t+1} \mid C_t, \cdots, C_1) = P(C_{t+1} \mid C_t)$$

Ïù¥ Îïå ÌäπÏ†ï ÏãúÍ∞Ñ  ùë°  ÎèôÏïà ÌäπÏ†ïÌïú Ìïú ÏÉÅÌÉú  ùëñ  ÏóêÏÑú ÌäπÏ†ïÌïú Îã§Î•∏ ÏÉÅÌÉú  ùëó  Î°ú Ï†ÑÏù¥Ìï† ÌôïÎ•†ÏùÑ Ï†ÑÏù¥ ÌôïÎ•†(Transition Probability) Ïù¥ÎùºÍ≥† ÌïúÎã§.

$$\gamma_{ij}(t) = P(C_{s+t}=j|C_{s}=i)$$

ÎòêÌïú Î™®Îì† ÏÉÅÌÉú Ï°∞Ìï©Ïóê ÎåÄÌïú Ï†ÑÏù¥ ÌôïÎ•†ÏùÑ ÎÇòÌÉÄÎÇ∏ Í≤ÉÏù¥ Ï†ÑÏù¥ ÌôïÎ•† ÌñâÎ†¨(Transition Probability Matirx)Ïù¥Îã§.

$$\Gamma(t) = \{ \gamma_{ij}(t) \}, \;\;\; \Gamma = \Gamma(1)$$

Ï≤¥ÌîÑÎ®º-ÏΩúÎ™®Í≥†Î°úÌîÑ Î∞©Ï†ïÏãù(Chapman-Kolmogorov EquationÏóê ÏùòÌïòÎ©¥ ÏãúÍ∞Ñ  ùë°+ùë¢  Ïùò Ï†ÑÏù¥ÌôïÎ•†ÌñâÎ†¨ÏùÄ ÏãúÍ∞Ñ  ùë°  Ïùò Ï†ÑÏù¥ÌôïÎ•†ÌñâÎ†¨Í≥º ÏãúÍ∞Ñ  ùë¢  Ïùò Ï†ÑÏù¥ÌôïÎ•†ÌñâÎ†¨Ïùò Í≥±ÏúºÎ°ú ÎÇòÌÉÄÎÇúÎã§.

$$\Gamma(t+u) = \Gamma(t)\Gamma(u)$$

## ÌûàÎì† ÎßàÏΩîÌîÑ Î™®Ìòï(Hidden Markov Model)

#### 1. ÌûàÎì† ÎßàÏΩîÌîÑ Î™®Ìòï(Hidden Markov Model)ÏùÄ ÎèÖÎ¶Ω ÌòºÌï© Î™®ÌòïÏóêÏÑú Ïó∞ÏÜçÌôïÎ•† Î∂ÑÌè¨Î•º ÏÑ†ÌÉùÌïòÎäî Ïù¥ÏÇ∞ÌôïÎ•† Í≥ºÏ†ï  ùê∂_ùë° Í∞Ä ÎßàÏΩîÌîÑ Ï≤¥Ïù∏Ïù¥Í≥† 

#### 2. Ïó∞ÏÜçÌôïÎ•† Î∂ÑÌè¨ ùëã_ùë°  Í∞Ä Í∑∏ ÏãúÏ†êÏùò Ïù¥ÏÇ∞ÌôïÎ•† Í≥ºÏ†ïÏùò Í∞íÏóêÎßå ÏùòÏ°¥ÌïòÎäî Î™®ÌòïÏùÑ ÎßêÌïúÎã§. 

Í∑∏Îü¨ÎÇò Ïó∞ÏÜç ÌôïÎ•† Î∂ÑÌè¨Ïùò Í∞í  ùëã_ùë°  Îßå Ï∏°Ï†ï Í∞ÄÎä•ÌïòÍ≥† Ïù¥ÏÇ∞ ÌôïÎ•† Í≥ºÏ†ïÏùò Í∞í  ùê∂_ùë°  Îäî Ï∏°Ï†ïÌï† Ïàò ÏóÜÎã§.

$$P(C_t \mid |C_{t-1}, \cdots, C_1) = P(C_t \mid C_{t-1})$$

$$P(X_t \mid X_t, \cdots, X_1, C_t, \cdots, C_1) = P(X_t \mid C_t)$$

## hmmlearn Ìå®ÌÇ§ÏßÄ

PythonÏóêÎäî Îã§ÏñëÌïú HMM ÏãúÎÆ¨Î†àÏù¥ÏÖò Î∞è Ï∂îÏ†ïÏö© Ìå®ÌÇ§ÏßÄÍ∞Ä Ï†úÍ≥µÎêúÎã§. Ïó¨Í∏∞ÏóêÏÑúÎäî Í∑∏ Ï§ë ÌïòÎÇòÏù∏ hmmlearn Ìå®ÌÇ§ÏßÄÎ•º ÏÇ¨Ïö©Ìï¥Î≥¥Ïûê.

hmmlearnÏùò hmm ÏÑúÎ∏å Ìå®ÌÇ§ÏßÄÏóêÏÑúÎäî Îã§ÏùåÍ≥º Í∞ôÏùÄ ÏÑ∏Í∞ÄÏßÄ HMM Î™®Ìòï ÌÅ¥ÎûòÏä§Î•º Ï†úÍ≥µÌïúÎã§.

+ GaussianHMM
    + Hidden Markov Model with Gaussian emissions.
+ GMMHMM
    + Hidden Markov Model with Gaussian mixture emissions.
+ MultinomialHMM
    + Hidden Markov Model with multinomial (discrete) emissions

GaussianHMM ÌÅ¥ÎûòÏä§Î•º ÏÇ¨Ïö©ÌïòÎ©¥ Ïπ¥ÌÖåÍ≥†Î¶¨-Îã§Î≥ÄÏàò Ï†ïÍ∑úÎ∂ÑÌè¨ ÌòºÌï© Î™®ÌòïÏùÑ ÏãúÎÆ¨Î†àÏù¥ÏÖò Ìï† Ïàò ÏûàÎã§.

Ïù¥ Îïå Îã§ÏùåÍ≥º Í∞ôÏùÄ ÏÜçÏÑ±ÏùÑ ÏÑ§Ï†ïÌï† Ïàò ÏûàÎã§.

+ startprob_: Ï¥àÍ∏∞ ÌôïÎ•† Î≤°ÌÑ∞
+ transmat_: Ï†ÑÏù¥ ÌôïÎ•† ÌñâÎ†¨
+ means_: Ï†ïÍ∑ú Î∂ÑÌè¨Ïùò Í∏∞ÎåìÍ∞í Î≤°ÌÑ∞
+ covars_: Ï†ïÍ∑ú Î∂ÑÌè¨Ïùò Í≥µÎ∂ÑÏÇ∞ ÌñâÎ†¨

Í∞ÄÏû• Í∞ÑÎã®Ìïú Î≤†Î•¥ÎàÑÏù¥-Ï†ïÍ∑ú Î∂ÑÌè¨Î•º ÏãúÌóòÌï¥Î≥¥Ïûê.

```python
import warnings
warnings.filterwarnings(action='ignore')

import matplotlib.pylab as plt
import matplotlib as mpl
import matplotlib.font_manager as fm
import seaborn as sns
import numpy as np

sns.set_style("whitegrid")
mpl.rcParams['axes.unicode_minus'] = False
plt.rcParams['font.size'] = 12

path = "/Library/Fonts/NanumGothic.otf"
font_name = fm.FontProperties(fname=path, size=20).get_name()

plt.rc('font', family=font_name)
```

```python
from hmmlearn import hmm
np.random.seed(3)

model = hmm.GaussianHMM(n_components=2, covariance_type='diag')
model.startprob_ = np.array([0.9, 0.1]) # Ï¥àÍ∏∞ ÌôïÎ•† Î≤°ÌÑ∞
model.transmat_ = np.array([[0.95, 0.05], [0.15, 0.85]]) # Ï†ÑÏù¥ÌôïÎ•† Î≤°ÌÑ∞
model.means_ = np.array([[1.0], [-3.0]]) # Ï†ïÍ∑úÎ∂ÑÌè¨ Í∏∞ÎåÄÍ∞í Î≤°ÌÑ∞
model.covars_ = np.array([[15.0], [40.0]]) # Ï†ïÍ∑úÎ∂ÑÌè¨ Í≥µÎ∂ÑÏÇ∞ ÌñâÎ†¨
X, Z = model.sample(500)
```

```python
plt.subplot(311)
plt.plot(X)
plt.title("random variable")
plt.subplot(312)
plt.plot(Z)
plt.title("discrete state")
plt.subplot(313)
plt.plot((1 + 0.01*X).cumprod())
plt.title("X cumulated")
plt.tight_layout()
plt.show()
```

```python
sns.distplot(X)
plt.show()
```

## ÎîîÏΩîÎìú

#### Í¥ÄÏ∏°Îêú ÌûàÎì† ÎßàÏΩîÌîÑ Î™®ÌòïÏùò Ïó∞ÏÜçÌôïÎ•† Î≥ÄÏàò Í∞íÏúºÎ°úÎ∂ÄÌÑ∞ ÎÇ¥Î∂ÄÏùò Ïù¥ÏÇ∞ ÌôïÎ•† Î≥ÄÏàò Í∞íÏùÑ Ï∂îÏ†ïÌïòÎäî Í≥ºÏ†ïÏùÑ ÎîîÏΩîÎìú(decode)ÎùºÍ≥† ÌïúÎã§. 

ÎîîÏΩîÎìú ÏïåÍ≥†Î¶¨Ï¶ò Ï§ë Í∞ÄÏû• ÎßéÏù¥ ÏÇ¨Ïö©ÎêòÎäî Í≤ÉÏùÄ Viterbi ÏïåÍ≥†Î¶¨Ï¶òÏù¥Îã§.

#### hmmlearn Ìå®ÌÇ§ÏßÄÏùò HMM ÌÅ¥ÎûòÏä§Îì§ÏùÄ Î™®Ìòï Ï∂îÏ†ïÏùÑ ÏúÑÌïú fit Î©îÏÑúÎìúÏôÄ ÎîîÏΩîÎî©ÏùÑ ÏúÑÌïú decode Î©îÏÑúÎìúÎ•º Ï†úÍ≥µÌïúÎã§. 

Îã§ÏùåÏùÄ ÏúÑÏóêÏÑú ÏãúÎÆ¨Î†àÏù¥ÏÖòÌïú Í≤∞Í≥ºÎ•º ÎîîÏΩîÎî©Ìïú ÏòàÏù¥Îã§.

```python
model2 = hmm.GaussianHMM(n_components=2, n_iter=len(X)).fit(X)
model2
```

```python
Z_hat = model2.decode(X)
Z_hat
```

```python
Z_hat = Z_hat[1]
X_cum = (1 + 0.01*X).cumprod()  # ÎàÑÏ†ÅÌôïÎ•†
X_cum_hat = X_cum.copy()
X_cum_hat[:5]
```

```python
X_cum_hat[Z_hat == 0] = np.nan  # Z_hat Í∞íÏù¥ 0Ïù∏ Ïù∏Îç±Ïä§Ïóê ÎåÄÌï¥ÏÑúÎäî Í∞í Ï†úÍ±∞
```

```python
# ÏúÑÏùò Í≤∞Í≥ºÎ•º ÌîåÎ°úÌåÖÌïòÎ©¥
plt.figure(figsize=(8, 6))
plt.subplot(211)
plt.plot(X_cum, lw=5)
plt.plot(X_cum_hat, color="red", lw=5)  # Z_hatÏù¥ 1Ïù∏ Î∂ÄÎ∂ÑÏùÑ Îπ®Í∞ÑÏÉâÏúºÎ°ú ÌëúÏãú
plt.title("X cumulated")
plt.subplot(212)
plt.plot(Z, 'bo-')
plt.plot(Z_hat, 'ro-')
plt.title("discrete state")
plt.tight_layout()
plt.show()
```

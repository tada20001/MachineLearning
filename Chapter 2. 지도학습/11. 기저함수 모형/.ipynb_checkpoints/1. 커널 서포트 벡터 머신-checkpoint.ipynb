## XOR Î¨∏Ï†ú

ÌçºÏÖâÌä∏Î°†Ïù¥ÎÇò ÏÑúÌè¨Ìä∏ Î≤°ÌÑ∞ Î®∏Ïã†Í≥º Í∞ôÏùÄ ÏÑ†ÌòïÌåêÎ≥ÑÌï®Ïàò Î∂ÑÎ•òÎ™®ÌòïÏùÄ Îã§ÏùåÍ≥º Í∞ôÏùÄ XOR(exclusive OR) Î¨∏Ï†úÎ•º ÌíÄÏßÄ Î™ªÌïúÎã§Îäî Îã®Ï†êÏù¥ ÏûàÎã§.

Ïù¥Îü¨Ìïú Í≤ΩÏö∞ÏóêÎäî Îã§Ïùå Í∑∏Î¶ºÏóêÏÑú Î≥¥ÎìØÏù¥ ÏÑ†ÌòïÌåêÎ≥ÑÌèâÎ©¥(decision hyperplane)ÏúºÎ°ú ÏòÅÏó≠ÏùÑ ÎÇòÎàå Ïàò ÏóÜÍ∏∞ ÎïåÎ¨∏Ïù¥Îã§.

```python
import warnings
warnings.filterwarnings(action='ignore')

import matplotlib.pylab as plt
import matplotlib as mpl
import matplotlib.font_manager as fm
import seaborn as sns
import numpy as np

sns.set_style("whitegrid")
mpl.rcParams['axes.unicode_minus'] = False
plt.rcParams['font.size'] = 12

path = "/Library/Fonts/NanumGothic.otf"
font_name = fm.FontProperties(fname=path, size=20).get_name()

plt.rc('font', family=font_name)
```

```python
np.random.seed(0)
X_xor = np.random.randn(200, 2)
y_xor = np.logical_xor(X_xor[:, 0] > 0, X_xor[:, 1] > 0)  # x1 > 0 & X2 > 0 False, x1 < 0 & x2 < 0 False
y_xor = np.where(y_xor, 1, 0)  # y_xor TrueÏù¥Î©¥ 1, ÏïÑÎãàÎ©¥ 0
plt.scatter(X_xor[y_xor==1, 0], X_xor[y_xor==1, 1], c='b', marker='o', label="ÌÅ¥ÎûòÏä§ 1", s=50)
plt.scatter(X_xor[y_xor==0, 0], X_xor[y_xor==0, 1], c='r', marker='s', label="ÌÅ¥ÎûòÏä§ 0", s=50)
plt.legend(loc=1)
plt.xlabel('x1')
plt.ylabel('x2')
plt.title("XOR Î¨∏Ï†ú")
plt.show()
```

ÏÑ†Ìòï SVMÏùÑ ÏÇ¨Ïö©ÌïòÎ©¥ XORÎ¨∏Ï†úÎ•º ÌíÄ Ïàò ÏóÜÎã§.

```python
def plot_xor(X, y, model, title, xmin=-3, xmax=3, ymin=-3, ymax=3):
    XX, YY = np.meshgrid(np.arange(xmin, xmax, (xmax-xmin)/1000), np.arange(ymin, ymax, (ymax-ymin)/1000))
    ZZ = model.predict(np.array([XX.ravel(), YY.ravel()]).T).reshape(XX.shape)
    plt.contourf(XX, YY, ZZ, cmap=mpl.cm.Paired_r, alpha=0.5)
    plt.scatter(X[y==1, 0], X[y==1, 1], c='b', marker='o', label='ÌÅ¥ÎûòÏä§ 1', s=50)
    plt.scatter(X[y==0, 0], X[y==0, 1], c='r', marker='s', label='ÌÅ¥ÎûòÏä§ 0', s=50)
    plt.xlim(xmin, xmax)
    plt.ylim(ymin, ymax)
    plt.title(title)
    plt.xlabel("x1")
    plt.ylabel("x2")
    
from sklearn.svm import SVC
svc = SVC(kernel='linear').fit(X_xor, y_xor)
plot_xor(X_xor, y_xor, svc, "ÏÑ†Ìòï SVC Î™®ÌòïÏùÑ ÏÇ¨Ïö©Ìïú XOR Î¨∏Ï†ú Î∂ÑÎ•ò Í≤∞Í≥º")
```

## Í∏∞Ï†ÄÌï®ÏàòÎ•º ÏÇ¨Ïö©Ìïú ÎπÑÏÑ†Ìòï ÌåêÎ≥Ñ Î™®Ìòï

Ïù¥Îü¨Ìïú Í≤ΩÏö∞ ÎèÑÏõÄÏù¥ ÎêòÎäî Í≤ÉÏù¥ ÏõêÎûòÏùò  ùê∑  Ï∞®Ïõê ÎèÖÎ¶Ω Î≥ÄÏàò Î≤°ÌÑ∞  ùë•  ÎåÄÏã†,
#### Í∏∞Ï†ÄÌï®Ïàò(basis function)ÏúºÎ°ú Î≥ÄÌôòÌïú  ùëÄ Ï∞®Ïõê Î≤°ÌÑ∞  ùúô(ùë•) Î•º ÎèÖÎ¶Ω Î≥ÄÏàò Î≤°ÌÑ∞Î°ú ÏÇ¨Ïö©ÌïòÎäî Î∞©Î≤ïÏù¥Îã§.

$$\phi(\cdot): {R}^D \rightarrow {R}^M$$

$$x=(x_1, x_2, \cdots, x_D) \;\;\; \rightarrow \;\;\; \phi(x) = (\phi_1(x), \phi_2(x), \cdots, \phi_M(x))$$

ÏïûÏÑú XOR Î¨∏Ï†úÎ•º ÌíÄÍ∏∞ ÏúÑÌï¥ Îã§ÏùåÍ≥º Í∞ôÏù¥ ÏÉÅÌò∏ Í≥±(cross-multiplication) Ìï≠ÏùÑ Ï∂îÍ∞ÄÌïú Í∏∞Ï†ÄÌï®ÏàòÎ•º ÏÇ¨Ïö©Ìï¥ Î≥¥Ïûê.

$$(x_1, x_2) \;\;\; \rightarrow \;\;\; \phi(x) = (x_1^2, \sqrt{2}x_1x_2, x_2^2)$$

```python
X = np.arange(6).reshape(3, 2)
X
```

FunctionTransformer Ï†ÑÏ≤òÎ¶¨ ÌÅ¥ÎûòÏä§Î°ú ÏúÑÏôÄ Í∏∞Ï†ÄÌï®ÏàòÎ•º Ïù¥Ïö©Ìïú ÌäπÏßï Î≥ÄÌôòÏùÑ Ìï† Ïàò ÏûàÎã§.

```python
from sklearn.preprocessing import FunctionTransformer

def basis(X):
    return np.vstack([X[:, 0] ** 2, np.sqrt(2) * X[:, 0] * X[:, 1], X[:, 1]**2]).T

FunctionTransformer(basis).fit_transform(X)
```

ÏúÑÏôÄ Í∞ôÏùÄ Í∏∞Ï†ÄÌï®ÏàòÎ•º Ïç®ÏÑú XOR Î¨∏Ï†úÏùò Îç∞Ïù¥ÌÑ∞Î•º Î≥ÄÌôòÌïòÎ©¥ ÌäπÏÑ±  ùúô2 Î•º ÏÇ¨Ïö©ÌïòÏó¨ ÌÅ¥ÎûòÏä§ Î∂ÑÎ•òÎ•º Ìï† Ïàò ÏûàÎã§Îäî Í≤ÉÏùÑ Ïïå Ïàò ÏûàÎã§.

```python
np.random.seed(0)
X_xor = np.random.randn(200, 2)
y_xor = np.logical_xor(X_xor[:, 0] > 0, X_xor[:, 1] > 0)  # x1 > 0 & X2 > 0 False, x1 < 0 & x2 < 0 False
y_xor = np.where(y_xor, 1, 0)  # y_xor TrueÏù¥Î©¥ 1, ÏïÑÎãàÎ©¥ 0

X_xor2 = FunctionTransformer(basis).fit_transform(X_xor)

plt.scatter(X_xor2[y_xor==1, 0], X_xor2[y_xor==1, 1], c='b', marker='o', s=50, label="ÌÅ¥ÎûòÏä§ 1")
plt.scatter(X_xor2[y_xor==0, 0], X_xor2[y_xor==0, 1], c='r', marker='s', s=50, label="ÌÅ¥ÎûòÏä§ 0")
plt.ylim(-6, 6)
plt.title("Î≥ÄÌôò Í≥µÍ∞ÑÏóêÏÑúÏùò Îç∞Ïù¥ÌÑ∞ Î∂ÑÌè¨")
plt.xlabel("$\phi_1$")
plt.ylabel("$\phi_2$")
plt.legend()
plt.show()
```

Îã§Ïùå ÏΩîÎìúÎäî Pipeline ÌÅ¥ÎûòÏä§Î°ú Í∏∞Ï†ÄÌï®Ïàò Ï†ÑÏπòÎ¶¨Í∏∞ÏôÄ SVC ÌÅ¥ÎûòÏä§Î•º Ìï©Ïπú Î™®ÌòïÏùò Î∂ÑÎ•ò Í≤∞Í≥ºÏù¥Îã§.

```python
from sklearn.pipeline import Pipeline

basismodel = Pipeline([
    ('basis', FunctionTransformer(basis)),
    ('svc', SVC(kernel='linear'))]).fit(X_xor, y_xor)

plot_xor(X_xor, y_xor, basismodel, "Í∏∞Ï†ÄÌï®Ïàò SVC Î™®ÌòïÏùÑ ÏÇ¨Ïö©Ìïú XOR Î∂ÑÎ•ò Í≤∞Í≥º")
```

## Ïª§ÎÑê Ìä∏Î¶≠

ÏÑúÌè¨Ìä∏ Î≤°ÌÑ∞ Î®∏Ïã†Ïùò Í≤ΩÏö∞ Î™©Ï†Å Ìï®ÏàòÏôÄ ÏòàÏ∏° Î™®ÌòïÏùÄ Îã§ÏùåÍ≥º Í∞ôÏùÄ dual formÏúºÎ°ú ÌëúÌòÑÌï† Ïàò ÏûàÎã§.

$$L =  \sum_{n=1}^N a_n - \dfrac{1}{2}\sum_{n=1}^N\sum_{m=1}^N a_n a_m y_n y_m x_n^T x_m$$

$$y = w^T x - w_0 = \sum_{n=1}^N a_n y_n x_n^T x - w_0$$


Ïù¥ ÏàòÏãùÏóêÏÑú  ùë• Î•º Í∏∞Ï†ÄÌï®Ïàò Î≥ÄÌôòÏúºÎ°ú  ùúô(ùë•) Î°ú Î∞îÍæ∏Î©¥

$$L =  \sum_{n=1}^N a_n - \dfrac{1}{2}\sum_{n=1}^N\sum_{m=1}^N a_n a_m y_n y_m \phi(x_n)^T \phi(x_m)$$

$$y = w^T x - w_0 = \sum_{n=1}^N a_n y_n \phi(x_n)^T \phi(x) - w_0$$

Ïù¥ ÎêúÎã§. Ï¶â Î™®Îì† Í∏∞Ï†ÄÌï®ÏàòÎäî  ùúô(ùë•ùëñ)ùëáùúô(ùë•ùëó) Ïùò ÌòïÌÉúÎ°úÎßå ÏÇ¨Ïö©ÎêòÎ©∞ ÎèÖÎ¶ΩÏ†ÅÏúºÎ°ú ÏÇ¨Ïö©ÎêòÏßÄ ÏïäÎäîÎã§.

Îî∞ÎùºÏÑú Îëê Í∞úÏùò Î≥ÄÌôòÎêú ÎèÖÎ¶Ω Î≥ÄÏàò Î≤°ÌÑ∞Î•º ÎÇ¥Ï†Å(inner product) Ìïú Í∞í  ùúô(ùë•_ùëñ)^ùëá * ùúô(ùë•_ùëó) Î•º ÌïòÎÇòÏùò Ìï®ÏàòÎ°ú ÎÇòÌÉÄÎÇº Ïàò ÏûàÎã§.

$$k(x_i, x_j) = \phi(x_i)^T \phi(x_j)$$

Ïù¥Îü¨Ìïú Ìï®ÏàòÎ•º Ïª§ÎÑê(kernel)Ïù¥ÎùºÍ≥† ÌïúÎã§.

ÎåÄÏùëÌïòÎäî Í∏∞Ï†ÄÌï®ÏàòÍ∞Ä Ï°¥Ïû¨Ìï† ÏàòÎßå ÏûàÎã§Î©¥ Í∏∞Ï†ÄÌï®ÏàòÎ•º Î®ºÏ†Ä Ï†ïÏùòÌïòÍ≥† Ïª§ÎÑêÏùÑ Ï†ïÏùòÌïòÎäî Í≤ÉÏù¥ ÏïÑÎãàÎùº Ïª§ÎÑêÏùÑ Î®ºÏ†Ä Ï†ïÏùòÌï¥ÎèÑ ÏÉÅÍ¥ÄÏóÜÎã§.

## Ïª§ÎÑêÏùò ÏùòÎØ∏

ÏÑúÌè¨Ìä∏ Î≤°ÌÑ∞ Î®∏Ïã†Ïùò Î™©Ï†Å Ìï®ÏàòÏôÄ ÏòàÏ∏° Î™®ÌòïÏùÄ Ïª§ÎÑêÏùÑ ÏÇ¨Ïö©ÌïòÏó¨ ÌëúÌòÑÌïòÎ©¥ Îã§ÏùåÍ≥º Í∞ôÎã§.

$$L =  \sum_{n=1}^N a_n - \dfrac{1}{2}\sum_{n=1}^N\sum_{m=1}^N a_n a_m y_n y_m k(x_n, x_m)$$

$$y = w^T x - w_0 = \sum_{n=1}^N a_n y_n k(x_n, x) - w_0$$

Ïª§ÎÑêÏùÑ ÏÇ¨Ïö©ÌïòÏßÄ ÏïäÎäî Í≤ΩÏö∞  ùëò(ùë•,ùë¶)=ùë•^ùëáùë¶ ÎùºÎäî Ï†êÏùÑ Í≥†Î†§ÌïòÎ©¥ Ïª§ÎÑêÏùÄ Îã§ÏùåÍ≥º Í∞ôÏùÄ ÌäπÏßïÏùÑ Î≥¥Ïù∏Îã§.

+ ùë• ÏôÄ  ùë¶ Í∞Ä ÎèôÏùºÌïú Î≤°ÌÑ∞Ïùº Îïå Í∞ÄÏû• ÌÅ¨Í≥†
+ Îëê Î≤°ÌÑ∞Í∞ÑÏùò Í±∞Î¶¨Í∞Ä Î©ÄÏñ¥Ïßà ÏàòÎ°ù ÏûëÏïÑÏßÑÎã§.

Ï¶â, Îëê ÌëúÎ≥∏ Îç∞Ïù¥ÌÑ∞ Í∞ÑÏùò Ïú†ÏÇ¨ÎèÑ(similarity)Î•º Ï∏°Ï†ïÌïòÎäî Í∏∞Ï§ÄÏúºÎ°ú Î≥º ÏàòÎèÑ ÏûàÎã§.

## Ïª§ÎÑê ÏÇ¨Ïö©Ïùò Ïû•Ï†ê

Ïª§ÎÑêÏùÑ ÏÇ¨Ïö©ÌïòÎ©¥ Î≤†Ïù¥ÏãúÏä§ Ìï®ÏàòÎ•º ÌïòÎÇòÏî© Ï†ïÏùòÌïòÎäî ÏàòÍ≥†Î•º Îçú Ïàò ÏûàÏùÑ ÎøêÎçîÎü¨ Î≥ÄÌôòÍ≥º ÎÇ¥Ï†ÅÏóê Îì§Ïñ¥Í∞ÄÎäî Í≥ÑÏÇ∞ÎüâÏù¥ Ï§ÑÏñ¥Îì†Îã§. ÏòàÎ•º Îì§Ïñ¥, Îã§ÏùåÍ≥º Í∞ôÏùÄ Í∏∞Ï†ÄÌï®ÏàòÏùò Í≤ΩÏö∞

$$\phi(x_i) = \phi([x_{i,1}, x_{i,2}]) = (x_{i,1}^2, \sqrt{2}x_{i,1}x_{i,2}, x_{i,2}^2)$$

Ïª§ÎÑê Î∞©Î≤ïÏùÑ Ïì∞ÏßÄ ÏïäÏùÑ Í≤ΩÏö∞Ïóê  ùúô(ùë•ùëñ)^ùëáùúô(ùë•ùëó) Î•º Í≥ÑÏÇ∞ÌïòÎ†§Î©¥  4+4+3=11  Î≤àÏùò Í≥±ÏÖàÏùÑ Ìï¥Ïïº ÌïúÎã§.

+ ùúô(ùë•1)  Í≥ÑÏÇ∞ : Í≥±ÏÖà 4Ìöå
+ ùúô(ùë•2)  Í≥ÑÏÇ∞ : Í≥±ÏÖà 4Ìöå
+ ÎÇ¥Ï†Å: Í≥±ÏÖà 3Ìöå

Í∑∏Îü∞Îç∞ Ïù¥ Í∏∞Ï†ÄÌï®ÏàòÎäî Îã§ÏùåÍ≥º Í∞ôÏùÄ Ïª§ÎÑêÎ°ú ÎåÄÏ≤¥Í∞ÄÎä•ÌïòÎã§.

\begin{eqnarray}
k(x_1, x_2) 
&=& (x_1^Tx_2)^2 \\
&=& (x_{1,1}x_{2,1} + x_{1,2}x_{2,2})^2 \\
&=& x_{1,1}^2x_{2,1}^2 + 2x_{1,1}x_{2,1}x_{1,2}x_{2,2} + x_{1,2}^2y_{2,2}^2 \\
&=& (x_{1,1}^2, \sqrt{2}x_{1,1}x_{1,2}, x_{1,2}^2)  (x_{2,1}^2, \sqrt{2}x_{2,1}x_{2,2}, x_{2,2}^2)^T \\
&=& \phi(x_1)^T \phi(x_2)
\end{eqnarray}

Ïª§ÎÑêÏùÑ ÏÇ¨Ïö©ÌïòÎ©¥  ùúô(ùë•1)^ùëáùúô(ùë•2)  ÏùÑ Í≥ÑÏÇ∞ÌïòÎäîÎç∞  2+1=3  Î≤àÏùò Í≥±ÏÖàÏù¥Î©¥ ÎêúÎã§.

+ ùë•^ùëá1ùë•2 : Í≥±ÏÖà 2Ìöå
+ Ï†úÍ≥±: Í≥±ÏÖà 1Ìöå

## Ïª§ÎÑêÏùò ÌôïÏû• ÏÉùÏÑ±

Ïñ¥Îñ§ Ìï®ÏàòÍ∞Ä Ïª§ÎÑêÌï®ÏàòÍ∞Ä ÎêúÎã§Îäî Í≤ÉÏùÑ Ï¶ùÎ™ÖÌïòÍ∏∞ ÏúÑÌï¥ÏÑúÎäî Í∏∞Ï†ÄÌï®ÏàòÎ•º ÌïòÎÇò ÌïòÎÇò Ï†ïÏùòÌï† ÌïÑÏöîÏóÜÏù¥ Í∏∞Ï†ÄÌï®ÏàòÏùò ÎÇ¥Ï†ÅÏúºÎ°ú ÌëúÌòÑÌï† Ïàò ÏûàÎã§Îäî Í≤ÉÎßå Ï¶ùÎ™ÖÌïòÎ©¥ ÎêúÎã§. ÌïòÏßÄÎßå Ïã§Ï†úÎ°úÎäî Îã§Ïùå Í∑úÏπôÏùÑ Ïù¥Ïö©ÌïòÎ©¥ Ïù¥ÎØ∏ ÎßåÎì§Ïñ¥ÏßÑ Ïª§ÎÑê  ùëò1(ùë•1,ùë•2) ,  ùëò2(ùë•1,ùë•2) Î°úÎ∂ÄÌÑ∞ ÏÉàÎ°úÏö¥ Ïª§ÎÑêÏùÑ ÏâΩÍ≤å ÎßåÎì§ Ïàò ÏûàÎã§.

(1) Ïª§ÎÑêÌï®ÏàòÎ•º ÏñëÏàòÎ∞∞Ìïú Ìï®ÏàòÎäî Ïª§ÎÑêÌï®ÏàòÏù¥Îã§.

$$k(x_1, x_2) = ck_1(x_1, x_2)\;\;(c > 0)$$

(2) Ïª§ÎÑêÌï®ÏàòÏóê ÏñëÏàòÏù∏ ÏÉÅÏàòÎ•º ÎçîÌïú Ìï®ÏàòÎäî Ïª§ÎÑêÌï®ÏàòÏù¥Îã§.

$$k(x_1, x_2) = k_1(x_1, x_2) + c\;\;(c > 0)$$

(3) Îëê Ïª§ÎÑêÌï®ÏàòÎ•º ÎçîÌïú Ìï®ÏàòÎäî Ïª§ÎÑêÌï®ÏàòÏù¥Îã§.

$$k(x_1, x_2) = k_1(x_1, x_2) + k_2(x_1, x_2)$$

(4) Îëê Ïª§ÎÑêÌï®ÏàòÎ•º Í≥±Ìïú Ìï®ÏàòÎäî Ïª§ÎÑêÌï®ÏàòÏù¥Îã§.

$$k(x_1, x_2) = k_1(x_1, x_2)k_2(x_1, x_2)$$

(5) Ïª§ÎÑêÌï®ÏàòÎ•º  ùë•‚â•0 ÏóêÏÑú Îã®Ï°∞Ï¶ùÍ∞Ä(monotonically increasing)ÌïòÎäî Ìï®ÏàòÏóê Ï†ÅÏö©ÌïòÎ©¥ Ïª§ÎÑêÌï®ÏàòÏù¥Îã§.

$$k(x_1, x_2) = (k_1(x_1, x_2))^n \;\; (n=1, 2, \cdots)$$

$$k(x_1, x_2) = \exp(k_1(x_1, x_2))$$

$$k(x_1, x_2) = \text{sigmoid}(k_1(x_1, x_2))$$

(6) ùë•1,ùë•2  Í∞ÅÍ∞ÅÏùò Ïª§ÎÑêÌï®ÏàòÍ∞íÏùò Í≥±ÎèÑ Ïª§ÎÑêÌï®ÏàòÏù¥Îã§.

$$k(x_1, x_2) = k_1(x_1, x_1)k_2(x_2, x_2)$$

## ÎßéÏù¥ ÏÇ¨Ïö©ÎêòÎäî Ïª§ÎÑê

Îã§ÏùåÍ≥º Í∞ôÏùÄ Ïª§ÎÑêÎì§Ïù¥ ÎßéÏù¥ ÏÇ¨Ïö©ÎêòÎäî Ïª§ÎÑêÎì§Ïù¥Îã§. Ïù¥ Ïª§ÎÑêÎì§ÏùÄ ÎåÄÎ∂ÄÎ∂Ñ Í∏∞Ï†ÄÌï®ÏàòÎ°ú Î≥ÄÌôòÌïòÏòÄÏùÑ Îïå Î¨¥ÌïúÎåÄÏùò Ï∞®ÏõêÏùÑ Í∞ÄÏßÄÎäî Í∏∞Ï†ÄÌï®ÏàòÍ∞Ä ÎêúÎã§. Îî∞ÎùºÏÑú ÎåÄÎ∂ÄÎ∂ÑÏùò ÎπÑÏÑ†ÌòïÏÑ±ÏùÑ Ï≤òÎ¶¨Ìï† Ïàò ÏûàÎã§. ÎπÑÍµêÎ•º ÏúÑÌï¥ ÏÑ†Ìòï ÏÑúÌè¨Ìä∏ Î≤°ÌÑ∞ Î®∏Ïã†Ïùò Í≤ΩÏö∞ÎèÑ Ï∂îÍ∞ÄÌïòÏòÄÎã§.

+ ÏÑ†Ìòï ÏÑúÌè¨Ìä∏ Î≤°ÌÑ∞ Î®∏Ïã†

$$k(x_1, x_2) = x_1^Tx_2$$

+ Îã§Ìï≠ Ïª§ÎÑê (Polynomial Kernel)

$$k(x_1, x_2) = (\gamma (x_1^Tx_2) + \theta)^d$$

+ RBF(Radial Basis Function) ÎòêÎäî Í∞ÄÏö∞ÏãúÏïà Ïª§ÎÑê(Gaussian Kernel)

$$k(x_1, x_2) = \exp \left( -\gamma ||x_1-x_2||^2 \right)$$

+ ÏãúÍ∑∏Î™®Ïù¥Îìú Ïª§ÎÑê (Sigmoid Kernel)

$$k(x_1, x_2) = \tanh(\gamma (x_1^Tx_2) + \theta)$$

ÏïûÏóêÏÑú ÏÇ¨Ïö©Ìïú Í∏∞Ï†ÄÌï®ÏàòÎäî  ùõæ=1,  ùúÉ=0,  ùëë=2  Ïù∏ Îã§Ìï≠ Ïª§ÎÑêÏûÑÏùÑ Ïïå Ïàò ÏûàÎã§.

## Îã§Ìï≠ Ïª§ÎÑê

Îã§Ìï≠ Ïª§ÎÑêÏùÄ Î≤°ÌÑ∞Ïùò ÎÇ¥Ï†ÅÏúºÎ°ú Ï†ïÏùòÎêú Ïª§ÎÑêÏùÑ ÌôïÏû•ÌïòÏó¨ ÎßåÎì† Ïª§ÎÑêÏù¥Îã§. Îã§Ìï≠ Ïª§ÎÑêÏù¥ Ïñ¥Îñ§ Í∏∞Ï†ÄÌï®ÏàòÎ°ú ÎêòÏñ¥ ÏûàÎäîÏßÄ ÏïåÏïÑÎ≥¥Ïûê.

Í∞ÑÎã®Ìïú Í≤ΩÏö∞Î°ú  ùõæ=1 ,  ùúÉ=1 ,  ùëë=4 Ïù¥Í≥†  ùë• Í∞Ä Ïä§ÏπºÎùºÏù∏ Í≤ΩÏö∞ÏóêÎäî

\begin{eqnarray}
k(x_1, x_2) 
&=& (x_1^Tx_2 + 1)^4 \\
&=& x_1^4x_2^4 + 4x_1^3x_2^3 + 6x_1^2x_2^2 + 4x_1x_2 + 1 \\
&=& (x_1^4, 2x_1^3, \sqrt{6}x_1, 2x_1, 1)^T (x_2^4, 2x_2^3, \sqrt{6}x_2, 2x_2, 1) \ \\
\end{eqnarray}

Ï≤òÎüº Í∏∞Ï†ÄÌï®ÏàòÏùò ÎÇ¥Ï†ÅÏù¥ ÎêúÎã§. Ï¶â, Í∏∞Ï†ÄÌï®ÏàòÎäî Îã§Ïùå 5Í∞úÍ∞Ä ÎêúÎã§.

\begin{eqnarray}
\phi_1(x) &=& x^4 \\
\phi_2(x) &=& 2x^3 \\
\phi_3(x) &=& \sqrt{6}x^2 \\
\phi_4(x) &=& 2x \\
\phi_5(x) &=& 1 \\
\end{eqnarray}

```python
x1 = 1
x2 = np.linspace(-3, 3, 100)

def poly4(x1, x2):
    return (x1 * x2 + 1) ** 4

plt.figure(figsize=(8, 4))
plt.subplot(121)
plt.plot(x2, poly4(x1, x2), ls='-')
plt.xlabel("x2")
plt.title("4Ï∞® Îã§Ìï≠Ïª§ÎÑêÏùò Ïòà")

plt.subplot(122)
plt.plot(x2, x2 ** 4)
plt.plot(x2, 2 * x2**3 )
plt.plot(x2, np.sqrt(6) * x2 ** 2)
plt.plot(x2, 2 * x2)
plt.plot(x2, np.ones_like(x2))
plt.xlabel("x2")
plt.title("4Ï∞® Îã§Ìï≠Ïª§ÎÑêÏùò Í∏∞Ï†ÄÌï®ÏàòÎì§")

plt.show()
```

## RBF Ïª§ÎÑê

RBF Ïª§ÎÑêÏùÄ Í∞ÄÏö∞ÏãúÏïà Ïª§ÎÑêÏù¥ÎùºÍ≥†ÎèÑ ÌïúÎã§. Î¨∏Ï†úÎ•º Í∞ÑÎã®ÌïòÍ≤å ÌïòÍ∏∞ ÏúÑÌï¥ Îã§ÏùåÍ≥º Í∞ôÏù¥ Í∞ÄÏ†ïÌïúÎã§.

$$\gamma=\frac{1}{2}$$

$$\|x_1\| = \|x_2\| = 1$$

Í∑∏Îü¨Î©¥ RBF Ïª§ÎÑêÏùÄ

\begin{eqnarray}
k(x_1, x_2) 
&=& \exp{\left(-\frac{||x_1 - x_2||^2}{2}\right)} \\
&=& \exp{\left(-\frac{x_1^Tx_1}{2} - \frac{x_2^Tx_2}{2} + 2x_1^Tx_2 \right)} \\
&=& \exp{\left(-\frac{x_1^Tx_1}{2}\right)}\exp{\left(-\frac{x_2^Tx_2}{2}\right)}\exp{(x_1^Tx_2)} \\
&=& C \exp{(x_1^Tx_2)} \\
&\approx& C \left( 1 + (x_1^Tx_2) + \dfrac{1}{2!}(x_1^Tx_2)^2 +  \dfrac{1}{3!}(x_1^Tx_2)^3 + \cdots \right) \\
\end{eqnarray}

ÏóêÏÑú Ï∞®ÏàòÍ∞Ä Î¨¥ÌïúÎåÄÏù∏ Îã§Ìï≠Ïª§ÎÑêÍ≥º Í∞ôÎã§.

```python
x1 = 0.0
x2 = np.linspace(-7, 7, 100)

def rbf(x1, x2, gamma):
    return np.exp(-gamma * np.abs(x2 - x1) ** 2)

plt.figure(figsize=(8, 4))
plt.subplot(121)
plt.plot(x2, rbf(x1, x2, 1), ls="-", label="gamma = 1")
plt.plot(x2, rbf(x1, x2, 0.5), ls=":", label="gamma = 0.5")
plt.plot(x2, rbf(x1, x2, 5), ls="--", label="gamma = 5")
plt.xlabel("x1- x2")
plt.xlim(-3, 7)
plt.legend()
plt.title("RBF Ïª§ÎÑê")

plt.subplot(122)
plt.plot(x2, rbf(-4, x2, 1))
plt.plot(x2, rbf(-2, x2, 1))
plt.plot(x2, rbf(0, x2, 1))
plt.plot(x2, rbf(2, x2, 1))
plt.plot(x2, rbf(4, x2, 1))
plt.xlabel("x2")
plt.title("RBF Ïª§ÎÑêÏùò Í∏∞Ï†ÄÌï®ÏàòÎì§")
plt.show()
```

## scikit-learnÏùò Ïª§ÎÑê SVM

scikit-learnÏùò SVM ÌÅ¥ÎûòÏä§Îäî kernel Ïù∏ÏàòÎ•º ÏßÄÏ†ïÌïòÏó¨ Ïª§ÎÑêÏùÑ ÏÑ§Ï†ïÌï† Ïàò ÏûàÎã§.

+ kernel = "linear": ÏÑ†Ìòï SVM.  ùëò(ùë•1,ùë•2)=ùë•ùëá1ùë•2 
+ kernel = "poly": Îã§Ìï≠ Ïª§ÎÑê.  ùëò(ùë•1,ùë•2)=(ùõæ(ùë•ùëá1ùë•2)+ùúÉ)ùëë 
    + gamma:  ùõæ 
    + coef0:  ùúÉ 
    + degree:  ùëë 
+ kernel = "rbf" ÎòêÎäî kernel = None: RBF Ïª§ÎÑê.  ùëò(ùë•1,  ùë•2)=exp(‚àíùõæ||ùë•1‚àíùë•2||2) 
    + gamma:  ùõæ 

+ kernel = "sigmoid": ÏãúÍ∑∏Î™®Ïù¥Îìú Ïª§ÎÑê.  ùëò(ùë•1,  ùë•2)=tanh(ùõæ(ùë•^ùëá1ùë•2)+ùúÉ) 
    + gamma:  ùõæ 
    + coef0:  ùúÉ

## Î∂ìÍΩÉ Î¨∏Ï†úÏóêÏùò ÏùëÏö©

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
iris = load_iris()
X = iris.data[:, [2, 3]]  # ÍΩÉÏûé Í∏∏Ïù¥/Ìè≠
y = iris.target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)

# scaling
sc = StandardScaler()
sc.fit(X_train)
X_train_std = sc.transform(X_train)
X_test_std = sc.transform(X_test)
X_combined_std = np.vstack([X_train_std, X_test_std])
y_combined = np.hstack([y_train, y_test])
X_combined_std.shape, y_combined.shape
```

```python
def plot_iris(X, y, model, title, xmin=-2.5, xmax=2.5, ymin=-2.5, ymax=2.5):
    XX, YY = np.meshgrid(np.arange(xmin, xmax, (xmax-xmin)/1000), np.arange(ymin, ymax, (ymax-ymin)/1000))
    ZZ = np.reshape(model.predict(np.array([XX.ravel(), YY.ravel()]).T), XX.shape)
    plt.contourf(XX, YY, ZZ, cmap=mpl.cm.Paired_r, alpha=0.5)
    plt.scatter(X[y==0, 0], X[y==0, 1], c='r', marker='^', label='0', s=100)
    plt.scatter(X[y==1, 0], X[y==1, 1], c='g', marker='o', label='1', s=100)
    plt.scatter(X[y==2, 0], X[y==2, 1], c='b', marker='s', label='2', s=100)
    plt.xlim(xmin, xmax)
    plt.ylim(ymin, ymax)
    plt.xlabel("ÍΩÉÏûé Í∏∏Ïù¥")
    plt.ylabel("ÍΩÉÏûé Ìè≠")
    plt.title(title)
    
model1 = SVC(kernel='linear').fit(X_test_std, y_test)
model2 = SVC(kernel='poly', random_state=0, gamma=10, C=1.0).fit(X_test_std, y_test)
model3 = SVC(kernel='rbf', random_state=0, gamma=1, C=1.0).fit(X_test_std, y_test)
```

```python
plt.figure(figsize=(8, 12))
plt.subplot(311)
plot_iris(X_test_std, y_test, model1, "ÏÑ†Ìòï SVC")
plt.subplot(312)
plot_iris(X_test_std, y_test, model2, "Îã§Ìï≠Ïª§ÎÑê SVC")
plt.subplot(313)
plot_iris(X_test_std, y_test, model3, "RBF Ïª§ÎÑê SVM")
plt.tight_layout()
plt.show()
```

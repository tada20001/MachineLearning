#### ÌçºÏÖâÌä∏Î°†ÏùÄ Í∞ÄÏû• Îã®ÏàúÌïòÍ≥† Îπ†Î•∏ ÌåêÎ≥Ñ Ìï®Ïàò Í∏∞Î∞ò Î∂ÑÎ•ò Î™®ÌòïÏù¥ÏßÄÎßå ÌåêÎ≥Ñ Í≤ΩÍ≥ÑÏÑ†(decision hyperplane)Ïù¥ Ïú†ÎãàÌÅ¨ÌïòÍ≤å Ï°¥Ïû¨ÌïòÏßÄ ÏïäÎäîÎã§Îäî ÌäπÏßïÏù¥ ÏûàÎã§. 

#### ÏÑúÌè¨Ìä∏ Î≤°ÌÑ∞ Î®∏Ïã†(SVM: support vector machine)ÏùÄ ÌçºÏÖâÌä∏Î°† Í∏∞Î∞òÏùò Î™®ÌòïÏóê Í∞ÄÏû• ÏïàÏ†ïÏ†ÅÏù∏ ÌåêÎ≥Ñ Í≤ΩÍ≥ÑÏÑ†ÏùÑ Ï∞æÍ∏∞ ÏúÑÌïú Ï†úÌïú Ï°∞Í±¥ÏùÑ Ï∂îÍ∞ÄÌïú Î™®ÌòïÏù¥ÎùºÍ≥† Î≥º Ïàò ÏûàÎã§.

[Í∑∏Î¶º] ÏÑúÌè¨Ìä∏ Î≤°ÌÑ∞Î®∏Ïã† : https://datascienceschool.net/upfiles/3e0a11567a154fc6a5e25010a68b5872.png

## ÏÑúÌè¨Ìä∏ÏôÄ ÎßàÏßÑ

Îã§ÏùåÍ≥º Í∞ôÏù¥  ùëÅ Í∞úÏùò ÌïôÏäµÏö© Îç∞Ïù¥ÌÑ∞Í∞Ä ÏûàÎã§Í≥† ÌïòÏûê.

$$(x_1, y_1), (x_2, y_2), \ldots, (x_i, y_i), \ldots, (x_N, y_N)$$

ÌåêÎ≥ÑÌï®ÏàòÎ™®ÌòïÏóêÏÑú  ùë¶ Îäî  +1 ,  ‚àí1  Îëê Í∞úÏùò Í∞íÏùÑ Í∞ÄÏßÑÎã§.

$$y = \begin{cases} +1 \\ -1 \end{cases}$$

ùë•  Îç∞Ïù¥ÌÑ∞ Ï§ëÏóêÏÑú  ùë¶ Í∞íÏù¥ +1Ïù∏ Îç∞Ïù¥ÌÑ∞Î•º  ùë•+ ,  ùë¶ Í∞íÏù¥ -1Ïù∏ Îç∞Ïù¥ÌÑ∞Î•º  ùë•‚àí ÎùºÍ≥† ÌïòÏûê.

ÌåêÎ≥ÑÌï®Ïàò Î™®ÌòïÏóêÏÑú ÏßÅÏÑ†Ïù∏ ÌåêÎ≥Ñ Ìï®Ïàò  ùëì(ùë•) Îäî Îã§ÏùåÍ≥º Í∞ôÏùÄ ÏàòÏãùÏúºÎ°ú ÎÇòÌÉÄÎÇº Ïàò ÏûàÎã§.

$$f(x) = w^Tx-w_0$$

ÌåêÎ≥ÑÌï®ÏàòÏùò Ï†ïÏùòÏóê Îî∞Îùº  ùë¶  Í∞íÏù¥  +1 Ïù∏ Í∑∏ Îç∞Ïù¥ÌÑ∞  ùë•+ Ïóê ÎåÄÌïú ÌåêÎ≥ÑÌï®Ïàò Í∞íÏùÄ ÏñëÏàòÍ∞Ä ÎêúÎã§.

$$f(x_+) = w^Tx_+ - w_0 > 0$$

Î∞òÎåÄÎ°ú  ùë¶  Í∞íÏù¥  ‚àí1 Ïù∏ Í∑∏ Îç∞Ïù¥ÌÑ∞  ùë•‚àí Ïóê ÎåÄÌïú ÌåêÎ≥ÑÌï®Ïàò Í∞íÏùÄ ÏùåÏàòÍ∞Ä ÎêúÎã§.

$$f(x_-) = w^Tx_- - w_0 < 0$$

#### ùë¶  Í∞íÏù¥  +1 Ïù∏ Îç∞Ïù¥ÌÑ∞ Ï§ëÏóêÏÑú ÌåêÎ≥ÑÌï®ÏàòÏùò Í∞íÏù¥ Í∞ÄÏû• ÏûëÏùÄ Îç∞Ïù¥ÌÑ∞Î•º  ùë•+ ÎùºÍ≥† ÌïòÍ≥†,  ùë¶  Í∞íÏù¥  ‚àí1 Ïù∏ Îç∞Ïù¥ÌÑ∞ Ï§ëÏóêÏÑú ÌåêÎ≥ÑÌï®ÏàòÏùò Í∞íÏù¥ Í∞ÄÏû• ÌÅ∞ Îç∞Ïù¥ÌÑ∞Î•º  ùë•‚àí ÎùºÍ≥† ÌïòÏûê. 

#### Ïù¥ Îç∞Ïù¥ÌÑ∞Îì§ÏùÄ Í∞ÅÍ∞ÅÏùò ÌÅ¥ÎûòÏä§Ïóê ÏÜçÌïú Îç∞Ïù¥ÌÑ∞ Ï§ëÏóêÏÑú Í∞ÄÏû• Í≤ΩÍ≥ÑÏÑ†Ïóê Í∞ÄÍπåÏù¥ Î∂ôÏñ¥ÏûàÎäî ÏµúÏ†ÑÎ∞©(most front)Ïùò Îç∞Ïù¥ÌÑ∞Îì§Ïù¥Îã§. Ïù¥Îü¨Ìïú Îç∞Ïù¥ÌÑ∞Î•º ÏÑúÌè¨Ìä∏(support) ÌòπÏùÄ ÏÑúÌè¨Ìä∏ Î≤°ÌÑ∞(support vector)ÎùºÍ≥† ÌïúÎã§. 

Î¨ºÎ°† Ïù¥ ÏÑúÌè¨Ìä∏Ïóê ÎåÄÌï¥ÏÑúÎèÑ Î∂ÄÌò∏ Ï°∞Í±¥ÏùÄ ÎßåÏ°±ÎêòÏñ¥Ïïº ÌïúÎã§.

$$f(x^+) = w^Tx^+ - w_0 > 0$$

$$f(x^-) = w^Tx^- - w_0 < 0$$

ÏÑúÌè¨Ìä∏Ïóê ÎåÄÌïú ÌåêÎ≥Ñ Ìï®ÏàòÏùò Í∞í  ùëì(ùë•+) ,  ùëì(ùë•‚àí)  Í∞íÏùÄ Î∂ÄÌò∏ Ï°∞Í±¥Îßå ÏßÄÌÇ§Î©¥ Ïñ¥Îñ§ Í∞íÏù¥ ÎêòÏñ¥ÎèÑ Í¥úÏ∞ÆÎã§. Îî∞ÎùºÏÑú Îã§ÏùåÍ≥º Í∞ôÏùÄ Ï°∞Í±¥ÏùÑ ÎßåÏ°±ÌïòÎèÑÎ°ù ÌåêÎ≥Ñ Ìï®ÏàòÎ•º Íµ¨ÌïúÎã§.

$$f(x^+) = w^T x^{+} - w_0 = +1$$

$$f(x^-) = w^T x^{-} - w_0 = -1$$

[Í∑∏Î¶º] ÏÑúÌè¨Ìä∏Î≤°ÌÑ∞Ïùò ÌåêÎ≥ÑÌï®Ïàò Í∞í : https://datascienceschool.net/upfiles/d109dea1e0fa40f9901dbbeff456230f.png

Ïù¥Î†áÍ≤å ÎêòÎ©¥ Î™®Îì†  ùë•+ ,  ùë•-  Îç∞Ïù¥ÌÑ∞Ïóê ÎåÄÌï¥ ÌåêÎ≥ÑÌï®ÏàòÏùò Í∞íÏùò Ï†àÎåÄÍ∞íÏù¥ 1Î≥¥Îã§ Ïª§ÏßÄÎØÄÎ°ú Îã§Ïùå Î∂ÄÎì±ÏãùÏù¥ ÏÑ±Î¶ΩÌïúÎã§.

$$w^Tx_+ - w_o \geq 1$$

$$w^Tx_- - w_o \leq -1$$

ÌåêÎ≥Ñ Í≤ΩÍ≥ÑÏÑ†  ùë§ùëáùë•‚àíùë§0=0  Í≥º Ï†ê  ùë•+ ,  ùë•‚àí  ÏÇ¨Ïù¥Ïùò Í±∞Î¶¨Îäî Îã§ÏùåÍ≥º Í∞ôÏù¥ Í≥ÑÏÇ∞Ìï† Ïàò ÏûàÎã§.

$$\dfrac{w^T x^{+} - w_0}{\| w \|} = \dfrac{1}{\| w \|}$$

$$-\dfrac{w^T x^{-} - w_0}{\| w \|} = \dfrac{1}{\| w \|}$$

ÎßàÏßÑ Í∞íÏù¥ ÏµúÎåÄÍ∞Ä ÎêòÎäî Í≤ΩÏö∞Îäî  ‚Äñùë§‚Äñ  Ï¶â,  ‚Äñùë§‚Äñ^2 Í∞Ä ÏµúÏÜåÍ∞Ä ÎêòÎäî Í≤ΩÏö∞ÏôÄ Í∞ôÎã§. Ï¶â Îã§ÏùåÍ≥º Í∞ôÏùÄ Î™©Ï†ÅÌï®ÏàòÎ•º ÏµúÏÜåÌôîÌïòÎ©¥ ÎêúÎã§.

$$L = \dfrac{1}{2} ||w||^2 = \dfrac{1}{2} w^T w$$

ÎòêÌïú Î™®Îì† ÌëúÎ≥∏ Îç∞Ïù¥ÌÑ∞Ïóê ÎåÄÌï¥ Î∂ÑÎ•òÎäî Ï†úÎåÄÎ°ú ÎêòÏñ¥Ïïº ÌïòÎØÄÎ°ú Î™®Îì† Îç∞Ïù¥ÌÑ∞  ùë•ùëñ, ùë¶ùëñ(ùëñ=1,‚Ä¶,ùëÅ) Ïóê ÎåÄÌï¥ Îã§Ïùå Ï°∞Í±¥ÏùÑ ÎßåÏ°±Ìï¥Ïïº ÌïúÎã§. ÏúÑÏóêÏÑú Ïä§ÏºÄÏùºÎßÅÏùÑ ÏÇ¨Ïö©ÌïòÏó¨ Î™®Îì† Îç∞Ïù¥ÌÑ∞Ïóê ÎåÄÌï¥  ùëì(ùë•ùëñ)=ùë§ùëáùë•ùëñ‚àíùë§ùëú  Í∞Ä 1Î≥¥Îã§ ÌÅ¨Í±∞ÎÇò -1Î≥¥Îã§ ÏûëÍ≤å ÎßåÎì§ÏóàÎã§Îäî Ï†êÏùÑ Ïù¥Ïö©ÌïúÎã§.

$$y_i \cdot f(x_i) = y_i \cdot( w^Tx_i - w_o) \geq 1 \;\;\; ( i = 1, \ldots, N )$$

$$y_i \cdot ( w^Tx_i - w_o) - 1 \geq 0 \;\;\; ( i = 1, \ldots, N )$$

ÎùºÍ∑∏ÎûëÏ£º ÏäπÏàòÎ≤ïÏùÑ ÏÇ¨Ïö©ÌïòÎ©¥ ÏµúÏÜåÌôî Î™©Ï†ÅÌï®ÏàòÎ•º Îã§ÏùåÍ≥º Í∞ôÏù¥ Í≥†ÏπòÎ©¥ ÎêúÎã§.

$$L = \dfrac{1}{2} w^T w - \sum_{i=1}^N a_i \{ y_i \cdot ( w^Tx_i - w_o) - 1 \}$$

Ïù¥ ÏµúÏ†ÅÌôî Î¨∏Ï†úÎ•º ÌíÄÏñ¥  ùë§ ,  ùë§0 ,  ùëé Î•º Íµ¨ÌïòÎ©¥ ÌåêÎ≥ÑÌï®ÏàòÎ•º ÏñªÏùÑ Ïàò ÏûàÎã§.

KKT(Karush‚ÄìKuhn‚ÄìTucker) Ï°∞Í±¥Ïóê Îî∞Î•¥Î©¥ Î∂ÄÎì±Ïãù Ï†úÌïú Ï°∞Í±¥Ïù¥ ÏûàÎäî Í≤ΩÏö∞ÏóêÎäî Îì±Ïãù Ï†úÌïúÏ°∞Í±¥ÏùÑ Í∞ÄÏßÄÎäî ÎùºÍ∑∏ÎûëÏ£º ÏäπÏàò Î∞©Î≤ïÍ≥º ÎπÑÏä∑ÌïòÏßÄÎßå  ùëñ Î≤àÏß∏ Î∂ÄÎì±ÏãùÏù¥ ÏûàÏúºÎÇò ÏóÜÏúºÎÇò ÎãµÏù¥ Í∞ôÏùÄ Í≤ΩÏö∞ÏóêÎäî ÎùºÍ∑∏ÎûëÏßÄ ÏäπÏàòÏùò Í∞íÏù¥  ùëéùëñ=0 Ïù¥ ÎêúÎã§. 

Ïù¥ Í≤ΩÏö∞Îäî ÌåêÎ≥ÑÌï®ÏàòÏùò Í∞í  ùë§ùëáùë•ùëñ‚àíùë§ùëú Ïù¥  ‚àí1 Î≥¥Îã§ ÏûëÍ±∞ÎÇò  1 Î≥¥Îã§ ÌÅ∞ Í≤ΩÏö∞Ïù¥Îã§.

$$y_i(w^Tx_i - w_o) - 1  > 0$$

#### ÌïôÏäµ Îç∞Ïù¥ÌÑ∞ Ï§ëÏóêÏÑú ÏµúÏ†ÑÎ∞© Îç∞Ïù¥ÌÑ∞Ïù∏ ÏÑúÌè¨Ìä∏ Î≤°ÌÑ∞Í∞Ä ÏïÑÎãå Î™®Îì† Îç∞Ïù¥ÌÑ∞Îì§Ïóê ÎåÄÌï¥ÏÑúÎäî Ïù¥ Ï°∞Í±¥Ïù¥ ÎßåÏ°±ÎêòÎØÄÎ°ú ÏÑúÌè¨Ìä∏ Î≤°ÌÑ∞Í∞Ä ÏïÑÎãå Îç∞Ïù¥ÌÑ∞Îäî ÎùºÍ∑∏ÎûëÏßÄ ÏäπÏàòÍ∞Ä 0Ïù¥ÎùºÎäî Í≤ÉÏùÑ Ïïå Ïàò ÏûàÎã§.

$$a_i = 0 \;\; \text{if} \;\; x_i \notin \{ x^{+}, x^{-} \}$$


## ÎìÄÏñº ÌòïÏãù

ÏµúÏ†ÅÌôî Ï°∞Í±¥ÏùÄ Î™©Ï†ÅÌï®Ïàò  ùêø ÏùÑ  ùë§ ,  ùë§0 Î°ú ÎØ∏Î∂ÑÌïú Í∞íÏù¥ 0Ïù¥ ÎêòÏñ¥Ïïº ÌïòÎäî Í≤ÉÏù¥Îã§.

$$\dfrac{\partial L}{\partial w} = 0$$

$$\dfrac{\partial L}{\partial w_0} = 0$$

Ïù¥ ÏãùÏùÑ ÌíÄÏñ¥ÏÑú Ï†ïÎ¶¨ÌïòÎ©¥ Îã§ÏùåÍ≥º Í∞ôÏïÑÏßÑÎã§.

\begin{eqnarray}
\dfrac{\partial L}{\partial w} 
&=& \dfrac{\partial}{\partial w} \left( \dfrac{1}{2} w^T w \right) - 
    \dfrac{\partial}{\partial w} \sum_{i=1}^N \left( a_i y_i w^Tx_i - a_i y_i w_o - a_i \right) \\
&=& w - \sum_{i=1}^N  a_i y_i x_i \\
&=& 0
\end{eqnarray}

\begin{eqnarray}
\dfrac{\partial L}{\partial w_0} 
&=& \dfrac{\partial}{\partial w_0} \left( \dfrac{1}{2} w^T w \right) - 
    \dfrac{\partial}{\partial w_0} \sum_{i=1}^N \left( a_i y_i w^Tx_i - a_i y_i w_o - a_i \right) \\
&=& \sum_{i=1}^N  a_i y_i \\
&=& 0
\end{eqnarray}

Ï¶â,

$$w = \sum_{i=1}^N a_i y_i x_i$$

$$0 = \sum_{i=1}^N a_i y_i$$

Ïù¥ Îëê ÏàòÏãùÏùÑ ÏõêÎûòÏùò Î™©Ï†ÅÌï®ÏàòÏóê ÎåÄÏûÖÌïòÏó¨  ùë§ ,  ùë§0 ÏùÑ ÏóÜÏï†Î©¥ Îã§ÏùåÍ≥º Í∞ôÎã§.

\begin{eqnarray}
L
&=& \dfrac{1}{2} w^T w - \sum_{i=1}^N a_i \{ y_i \cdot ( w^Tx_i - w_o) - 1 \}  \\
&=& \dfrac{1}{2} \left( \sum_{i=1}^N a_i y_i x_i \right)^T \left( \sum_{j=1}^N a_j y_j x_j \right) - 
    \sum_{i=1}^N a_i \left\{ y_i \cdot \left( \left( \sum_{j=1}^N a_j y_j x_j \right)^Tx_i - w_o \right) - 1 \right\}  \\
&=& \dfrac{1}{2} \sum_{i=1}^N \sum_{j=1}^N a_i a_j y_i y_j x_i^T x_j - 
    \sum_{i=1}^N \sum_{j=1}^N a_i a_j y_i y_j x_i^T x_j +
    w_0 \sum_{i=1}^N a_i y_i + \sum_{i=1}^N a_i   \\
&=& \sum_{i=1}^N a_i - \dfrac{1}{2}\sum_{i=1}^N\sum_{j=1}^N a_i a_j y_i y_j x_i^T x_j
\end{eqnarray}

Ï¶â,

$$L = \sum_{i=1}^N a_i - \dfrac{1}{2}\sum_{i=1}^N\sum_{j=1}^N a_i a_j y_i y_j x_i^T x_j$$

Ïù¥ Îïå  ùëé Îäî Îã§Ïùå Ï°∞Í±¥ÏùÑ ÎßåÏ°±ÌïúÎã§.

$$\sum_{i=1}^N a_i y_i = 0$$

$$a_i \geq 0 \;\;\;  ( i = 1, \ldots, N )$$

#### Ïù¥ Î¨∏Ï†úÎäî  ùë§ Î•º Íµ¨ÌïòÎäî Î¨∏Ï†úÍ∞Ä ÏïÑÎãàÎùº  ùëé ÎßåÏùÑ Íµ¨ÌïòÎäî Î¨∏Ï†úÎ°ú Î∞îÎÄåÏóàÏúºÎØÄÎ°ú ÎìÄÏñºÌòïÏãù(dual form)Ïù¥ÎùºÍ≥† ÌïúÎã§. 

ÎìÄÏñºÌòïÏãùÏúºÎ°ú Î∞îÍæ∏Î©¥ ÏàòÏπòÏ†ÅÏúºÎ°ú Î∞ïÏä§(Box)Ï†úÌïú Ï°∞Í±¥Ïù¥ ÏûàÎäî Ïù¥Ï∞®ÌîÑÎ°úÍ∑∏ÎûòÎ∞ç(QP; quadratic programming) Î¨∏Ï†úÍ∞Ä ÎêòÎØÄÎ°ú ÏõêÎûòÏùò Î¨∏Ï†úÎ≥¥Îã§Îäî Ìö®Ïú®Ï†ÅÏúºÎ°ú ÌíÄ Ïàò ÏûàÎã§.

ÎìÄÏñºÌòïÏãù Î¨∏Ï†úÎ•º ÌíÄÏñ¥ Ìï®Ïàò  ùêø  Î•º ÏµúÏÜåÌôîÌïòÎäî  ùëé Î•º Íµ¨ÌïòÎ©¥ ÏòàÏ∏° Î™®ÌòïÏùÑ Îã§ÏùåÍ≥º Í∞ôÏù¥ Ïì∏ Ïàò ÏûàÎã§.

$$f(x) = w^T x - w_0 = \sum_{i=1}^N a_i y_i x_i^T x - w_0$$

ùë§0 Îäî

$$w_0 = w^T x^{+} - 1$$

ÎòêÎäî

$$w_0 = w^T x^{-} + 1$$

ÎòêÎäî

$$w_0 = \dfrac{1}{2} w^T (x^+ + x^{-})$$

Î°ú Íµ¨ÌïúÎã§.

ÎùºÍ∑∏ÎûëÏ£º ÏäπÏàò Í∞íÏù¥ 0 Ï¶â,  ùëéùëñ=0  Ïù¥Î©¥ Ìï¥Îãπ Îç∞Ïù¥ÌÑ∞Îäî ÏòàÏ∏° Î™®Ìòï, Ï¶â  ùë§  Í≥ÑÏÇ∞Ïóê ÏïÑÎ¨¥Îü∞ Í∏∞Ïó¨Î•º ÌïòÏßÄ ÏïäÏúºÎØÄÎ°ú ÏúÑ ÏãùÏùÑ Ïã§Ï†úÎ°úÎäî Îã§ÏùåÍ≥º Í∞ôÎã§.

$$f(x) = a^+ x^T x^+ - a^- x^T x^- - w_0$$

#### Ïó¨Í∏∞ÏóêÏÑú  ùë•ùëáùë•+  Îäî  ùë• ÏôÄ  ùë•+ ÏÇ¨Ïù¥Ïùò  (ÏΩîÏÇ¨Ïù∏)Ïú†ÏÇ¨ÎèÑ,  ùë•ùëáùë•‚àí  Îäî  ùë• ÏôÄ  ùë•‚àí ÏÇ¨Ïù¥Ïùò  (ÏΩîÏÇ¨Ïù∏)Ïú†ÏÇ¨ÎèÑÏù¥ÎØÄÎ°ú Í≤∞Íµ≠ Îëê ÏÑúÌè¨Ìä∏ Î≤°ÌÑ∞ÏôÄÏùò Ïú†ÏÇ¨ÎèÑÎ•º Ï∏°Ï†ïÌï¥ÏÑú Í∞íÏù¥ ÌÅ∞ Ï™ΩÏúºÎ°ú ÌåêÎ≥ÑÌïòÍ≤å ÎêúÎã§.

## Scikit-LearnÏùò ÏÑúÌè¨Ìä∏ Î≤°ÌÑ∞ Î®∏Ïã†

Scikit-LearnÏùò svm ÏÑúÎ∏åÌéòÌÇ§ÏßÄÎäî ÏÑúÌè¨Ìä∏ Î≤°ÌÑ∞ Î®∏Ïã† Î™®ÌòïÏù∏ SVC (Support Vector Classifier) ÌÅ¥ÎûòÏä§Î•º Ï†úÍ≥µÌïúÎã§.

```python
import warnings
warnings.filterwarnings(action='ignore')

import matplotlib.pylab as plt
import matplotlib as mpl
import matplotlib.font_manager as fm
import seaborn as sns
import numpy as np

sns.set_style("whitegrid")
mpl.rcParams['axes.unicode_minus'] = False
plt.rcParams['font.size'] = 12

path = "/Library/Fonts/NanumGothic.otf"
font_name = fm.FontProperties(fname=path, size=20).get_name()

plt.rc('font', family=font_name)

from sklearn.svm import SVC
from sklearn.datasets import make_blobs
```

```python
X, y = make_blobs(n_samples=50, centers=2, cluster_std=0.5, random_state=4)
y = y * 2 - 1  # y Í∞íÏùÑ (1, -1)Î°ú Î≥ÄÌôò

plt.scatter(X[y==-1, 0], X[y==-1, 1], marker='o', label="-1 ÌÅ¥ÎûòÏä§")
plt.scatter(X[y==1, 0], X[y==1, 1], marker='x', label="1 ÌÅ¥ÎûòÏä§")
plt.xlabel("x1")
plt.ylabel("x2")
plt.legend()
plt.title("ÌïôÏäµÏö© Îç∞Ïù¥ÌÑ∞")
plt.show()
```

SVC ÌÅ¥ÎûòÏä§Îäî Ïª§ÎÑê(kernel)ÏùÑ ÏÑ†ÌÉùÌïòÎäî Ïù∏Ïàò kernelÍ≥º Ïä¨ÎûôÎ≥ÄÏàò Í∞ÄÏ§ëÏπò(slack variable weight)Î•º ÏÑ†ÌÉùÌïòÎäî Ïù∏Ïàò CÎ•º Î∞õÎäîÎç∞ ÏßÄÍ∏àÍπåÏßÄ Í≥µÎ∂ÄÌïú ÏÑúÌè¨Ìä∏ Î≤°ÌÑ∞ Î®∏Ïã†ÏùÑ ÏÇ¨Ïö©ÌïòÎ†§Î©¥ Ïù∏ÏàòÎ•º Îã§ÏùåÏ≤òÎüº ÎÑ£Ïñ¥Ï§ÄÎã§. Ïù¥ Ïù∏ÏàòÎì§Ïóê ÎåÄÌï¥ÏÑúÎäî Í≥ß ÏÑ§Î™ÖÌïúÎã§.

```python
from sklearn.svm import SVC
model = SVC(kernel='linear', C=1e10).fit(X, y)
```

SVCÎ•º ÏÇ¨Ïö©ÌïòÏó¨ Î™®ÌòïÏùÑ Íµ¨ÌïòÎ©¥ Îã§ÏùåÍ≥º Í∞ôÏùÄ ÏÜçÏÑ±Í∞íÏùÑ Í∞ÄÏßÑÎã§.

+ n_support_: Í∞Å ÌÅ¥ÎûòÏä§Ïùò ÏÑúÌè¨Ìä∏Ïùò Í∞úÏàò
+ support_: Í∞Å ÌÅ¥ÎûòÏä§Ïùò ÏÑúÌè¨Ìä∏Ïùò Ïù∏Îç±Ïä§
+ support_vectors_: Í∞Å ÌÅ¥ÎûòÏä§Ïùò ÏÑúÌè¨Ìä∏Ïùò x Í∞í.  ùë•+ ÏôÄ  ùë•‚àí 
+ coef_:  ùë§  Î≤°ÌÑ∞
+ intercept_:  ‚àíùë§0

## Ïä¨ÎûôÎ≥ÄÏàò

ÎßåÏïΩ Îç∞Ïù¥ÌÑ∞Í∞Ä ÏßÅÏÑ†Ïù∏ ÌåêÎ≥Ñ Í≤ΩÍ≥ÑÏÑ†ÏúºÎ°ú ÎÇòÎàÑÏñ¥ÏßÄÏßÄ ÏïäÎäî 

#### Ï¶â, ÏÑ†ÌòïÎ∂ÑÎ¶¨(linear separable)Í∞Ä Î∂àÍ∞ÄÎä•Ìïú Í≤ΩÏö∞ÏóêÎäî Îã§ÏùåÍ≥º Í∞ôÏù¥ Ïä¨ÎûôÎ≥ÄÏàò(slack variable)Î•º ÏÇ¨Ïö©ÌïòÏó¨ Í∞úÎ≥ÑÏ†ÅÏù∏ Ïò§Ï∞®Î•º ÌóàÏö©Ìï† Ïàò ÏûàÎã§.

ÏõêÎûò ÌåêÎ≥ÑÌï®ÏàòÏùò Í∞íÏùÄ ÌÅ¥ÎûòÏä§  +1  ÏòÅÏó≠Ïùò ÏÉòÌîå  ùë•+ Ïóê ÎåÄÌï¥

$$w^Tx_+ - w_0 \geq 1$$

ÌÅ¥ÎûòÏä§  ‚àí1  ÏòÅÏó≠Ïùò ÏÉòÌîå  ùë•‚àí Ïóê ÎåÄÌï¥

$$w^Tx_- - w_0 \leq -1$$

Ïù¥Ïñ¥Ïïº ÌïúÎã§.

ÏñëÏàòÏù∏ Ïä¨ÎûôÎ≥ÄÏàò  ùúâ‚â•0 Î•º ÏÇ¨Ïö©ÌïòÎ©¥ Ïù¥ Ï°∞Í±¥ÏùÑ Îã§ÏùåÍ≥º Í∞ôÏù¥ ÏôÑÌôîÌï† Ïàò ÏûàÎã§.

$$w^Tx_+ - w_0 \geq +1-\xi_i$$

Ïù¥ ÎêúÎã§.

Î™®Îì† Ïä¨ÎûôÎ≥ÄÏàòÎäî 0Î≥¥Îã§ Í∞ôÍ±∞ÎÇò ÌÅ¨Îã§.

$$\xi_i \geq 0 \;\;\; (i=1, \ldots, N)$$

ÏúÑÏùò Î∂ÄÎì±Ïãù Ï°∞Í±¥ÏùÑ Î™®Îëê Í≥†Î†§Ìïú ÏµúÏ†ÅÌôî Î™©Ï†ÅÌï®ÏàòÎäî Îã§ÏùåÍ≥º Í∞ôÏïÑÏßÑÎã§.

$$L = \dfrac{1}{2} ||w||^2 - \sum_{i=1}^N a_i (y_i \cdot ( w^Tx_i - w_o) - 1 + \xi_i ) - \sum_{i=1}^N \mu_i \xi_i  + C \sum_{i=1}^N \xi_i$$

ÏúÑ ÏãùÏóêÏÑú  ùê∂‚àëùëÅùëñ=1ùúâùëñ  Ìï≠(ÎßàÏßÄÎßâ Ìï≠)ÏùÄ Ïä¨ÎûôÎ≥ÄÏàòÏùò Ìï©Ïù¥ ÎÑàÎ¨¥ Ïª§ÏßÄÏßÄ ÏïäÎèÑÎ°ù Ï†úÌïúÌïòÎäî Ïó≠Ìï†ÏùÑ ÌïúÎã§.

[Í∑∏Î¶º] C Í∞íÏóê Îî∞Î•∏ ÌåêÎ≥ÑÌï®ÏàòÏùò Î≥ÄÌôî : https://datascienceschool.net/upfiles/ee724b3f4cf344278d0790eb8807630c.png

```python
np.random.seed(0)
X = np.r_[np.random.randn(20, 2) - [2, 2], np.random.randn(20, 2) + [2, 2]]
Y = [-1] * 20 + [1] * 20
X.shape, len(Y)
```

```python
plotnum = 1
for name, penalty in (('C=10', 10), ('C=0.1', 0.1)):
    clf = SVC(kernel='linear', C=penalty).fit(X, Y)
    xx = np.linspace(-5, 5)
    
    x_jin = -5
    x_jax = 5
    y_jin = -9
    y_jax = 9
    XX, YY = np.mgrid[x_jin:x_jax:200j, y_jin:y_jax:200j]
    
    levels = [-1, 0, 1]
    linestyles = ["dashed", "solid", "dashed"]
    Z = clf.decision_function(np.c_[XX.ravel(), YY.ravel()])
    Z = Z.reshape(XX.shape)
    
    plt.subplot(1, 2, plotnum)
    plt.contour(XX, YY, Z, levels, colors='k', linestyles=linestyles)
    plt.scatter(clf.support_vectors_[:, 0], clf.support_vectors_[:, 1], s=120, linewidth=4)
    plt.scatter(X[:, 0], X[:, 1], c=Y, s=30, lw=1, cmap=plt.cm.Paired)
    plt.xlim(x_jin, x_jax)
    plt.ylim(y_jin, y_jax)
    plt.title(name)
    
    plotnum += 1

plt.suptitle("Ïä¨ÎûôÎ≥ÄÏàò Í∞ÄÏ§ëÏπò CÏùò ÏòÅÌñ•", y=1.1)
plt.tight_layout()
plt.show()
```

```python
from sklearn.datasets import fetch_olivetti_faces
faces = fetch_olivetti_faces()

N = 2
M = 5
np.random.seed(0)
fig = plt.figure(figsize=(9, 5))
plt.subplots_adjust(top=1, bottom=0, hspace=0, wspace=0.05)
klist = np.random.choice(range(len(faces.data)), N*M)  # 10Í∞úÎßå ÏÑ†ÌÉù
for i in range(N):
    for j in range(M):
        k = klist[i*M+j]
        ax = fig.add_subplot(N, M, i*M+j+1)
        ax.imshow(faces.images[k], cmap=plt.cm.bone)
        ax.grid(False)
        ax.xaxis.set_ticks([])
        ax.yaxis.set_ticks([])
        plt.title(faces.target[k])
plt.tight_layout()
plt.show()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(faces.data, faces.target, test_size=0.4, random_state=0)

from sklearn.svm import SVC
svc = SVC(kernel='linear').fit(X_train, y_train)
N = 2
M = 5

np.random.seed(4)
fig = plt.figure(figsize=(9, 5))
plt.subplots_adjust(top=1, bottom=0, hspace=0, wspace=0.05)
klist = np.random.choice(range(len(y_test)), N*M)
for i in range(N):
    for j in range(M):
        k = klist[M * i + j]
        ax = fig.add_subplot(N, M, i * M + j + 1)
        ax.imshow(X_test[k:(k + 1), :].reshape(64, 64), cmap=plt.cm.bone)
        ax.grid(False)
        ax.xaxis.set_ticks([])
        ax.yaxis.set_ticks([])
        plt.title("%d => %d" % (y_test[k], svc.predict(X_test[k:(k+1), :])[0]))
        
plt.tight_layout()
plt.show()
```

```python
from sklearn.metrics import classification_report, accuracy_score

y_pred_train = svc.predict(X_train)
y_pred_test = svc.predict(X_test)
```

```python
print(classification_report(y_train, y_pred_train))
```

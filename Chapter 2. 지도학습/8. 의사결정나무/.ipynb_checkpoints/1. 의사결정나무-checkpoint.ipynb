#### ÏùòÏÇ¨Í≤∞Ï†ïÎÇòÎ¨¥(decision tree)Îäî Ïó¨Îü¨ Í∞ÄÏßÄ Í∑úÏπôÏùÑ ÏàúÏ∞®Ï†ÅÏúºÎ°ú Ï†ÅÏö©ÌïòÎ©¥ÏÑú ÎèÖÎ¶Ω Î≥ÄÏàò Í≥µÍ∞ÑÏùÑ Î∂ÑÌï†ÌïòÎäî Î∂ÑÎ•ò Î™®ÌòïÏù¥Îã§. 

Î∂ÑÎ•ò(classification)ÏôÄ ÌöåÍ∑Ä Î∂ÑÏÑù(regression)Ïóê Î™®Îëê ÏÇ¨Ïö©Îê† Ïàò ÏûàÍ∏∞ ÎïåÎ¨∏Ïóê CART(Classification And Regression Tree)ÎùºÍ≥†ÎèÑ ÌïúÎã§.

## ÏùòÏÇ¨Í≤∞Ï†ïÎÇòÎ¨¥Î•º Ïù¥Ïö©Ìïú Î∂ÑÎ•òÌïôÏäµ

ÏùòÏÇ¨Í≤∞Ï†ïÎÇòÎ¨¥Î•º Ïù¥Ïö©Ìïú Î∂ÑÎ•òÎ≤ïÏùÄ Îã§ÏùåÍ≥º Í∞ôÎã§.

(1) Ïó¨Îü¨Í∞ÄÏßÄ ÎèÖÎ¶Ω Î≥ÄÏàò Ï§ë ÌïòÎÇòÏùò ÎèÖÎ¶Ω Î≥ÄÏàòÎ•º ÏÑ†ÌÉùÌïòÍ≥† Í∑∏ ÎèÖÎ¶Ω Î≥ÄÏàòÏóê ÎåÄÌïú Í∏∞Ï§ÄÍ∞í(threshold)ÏùÑ Ï†ïÌïúÎã§. Ïù¥Î•º Î∂ÑÎ•ò Í∑úÏπôÏù¥ÎùºÍ≥† ÌïúÎã§. ÏµúÏ†ÅÏùò Î∂ÑÎ•ò Í∑úÏπôÏùÑ Ï∞æÎäî Î∞©Î≤ïÏùÄ Ïù¥ÌõÑÏóê ÏûêÏÑ∏Ìûà ÏÑ§Î™ÖÌïúÎã§.

(2) Ï†ÑÏ≤¥ ÌïôÏäµ Îç∞Ïù¥ÌÑ∞ ÏßëÌï©(Î∂ÄÎ™® ÎÖ∏Îìú)ÏùÑ Ìï¥Îãπ ÎèÖÎ¶Ω Î≥ÄÏàòÏùò Í∞íÏù¥ Í∏∞Ï§ÄÍ∞íÎ≥¥Îã§ ÏûëÏùÄ Îç∞Ïù¥ÌÑ∞ Í∑∏Î£π(ÏûêÏãù ÎÖ∏Îìú 1)Í≥º Ìï¥Îãπ ÎèÖÎ¶Ω Î≥ÄÏàòÏùò Í∞íÏù¥ Í∏∞Ï§ÄÍ∞íÎ≥¥Îã§ ÌÅ∞ Îç∞Ïù¥ÌÑ∞ Í∑∏Î£π(ÏûêÏãù ÎÖ∏Îìú 2)ÏúºÎ°ú ÎÇòÎààÎã§.

(3) Í∞ÅÍ∞ÅÏùò ÏûêÏãù ÎÖ∏ÎìúÏóê ÎåÄÌï¥ 1~2Ïùò Îã®Í≥ÑÎ•º Î∞òÎ≥µÌïòÏó¨ ÌïòÏúÑÏùò ÏûêÏãù ÎÖ∏ÎìúÎ•º ÎßåÎì†Îã§. Îã®, ÏûêÏãù ÎÖ∏ÎìúÏóê ÌïúÍ∞ÄÏßÄ ÌÅ¥ÎûòÏä§Ïùò Îç∞Ïù¥ÌÑ∞Îßå Ï°¥Ïû¨ÌïúÎã§Î©¥ Îçî Ïù¥ÏÉÅ ÏûêÏãù ÎÖ∏ÎìúÎ•º ÎÇòÎàÑÏßÄ ÏïäÍ≥† Ï§ëÏßÄÌïúÎã§.

Ïù¥Î†áÍ≤å ÏûêÏãù ÎÖ∏Îìú ÎÇòÎàÑÍ∏∞Î•º Ïó∞ÏÜçÏ†ÅÏúºÎ°ú Ï†ÅÏö©ÌïòÎ©¥ ÎÖ∏ÎìúÍ∞Ä Í≥ÑÏÜç Ï¶ùÍ∞ÄÌïòÎäî ÎÇòÎ¨¥(tree)ÏôÄ Í∞ôÏùÄ ÌòïÌÉúÎ°ú ÌëúÌòÑÌï† Ïàò ÏûàÎã§.


## ÏùòÏÇ¨Í≤∞Ï†ïÎÇòÎ¨¥Î•º ÏÇ¨Ïö©Ìïú Î∂ÑÎ•òÏòàÏ∏°

ÏùòÏÇ¨Í≤∞Ï†ïÎÇòÎ¨¥Ïóê Ï†ÑÏ≤¥ Ìä∏Î†àÏù¥Îãù Îç∞Ïù¥ÌÑ∞Î•º Î™®Îëê Ï†ÅÏö©Ìï¥ Î≥¥Î©¥ Í∞Å Îç∞Ïù¥ÌÑ∞Îäî ÌäπÏ†ïÌïú ÎÖ∏ÎìúÎ•º ÌÉÄÍ≥† ÎÇ¥Î†§Í∞ÄÍ≤å ÎêúÎã§. Í∞Å ÎÖ∏ÎìúÎäî Í∑∏ ÎÖ∏ÎìúÎ•º ÏÑ†ÌÉùÌïú Îç∞Ïù¥ÌÑ∞ ÏßëÌï©ÏùÑ Í∞ÄÏßÑÎã§. Ïù¥ Îïå ÎÖ∏ÎìúÏóê ÏÜçÌïú Îç∞Ïù¥ÌÑ∞Ïùò ÌÅ¥ÎûòÏä§Ïùò ÎπÑÏú®ÏùÑ Íµ¨ÌïòÏó¨ Ïù¥Î•º Í∑∏ ÎÖ∏ÎìúÏùò Ï°∞Í±¥Î∂Ä ÌôïÎ•† Î∂ÑÌè¨  ùëÉ(ùëå=ùëò|ùëã)node ÎùºÍ≥† Ï†ïÏùòÌïúÎã§.

$$P(Y=k|X)_{\text{node}} \approx \dfrac{N_{\text{node},k}}{N_{\text{node}}}$$

ÌÖåÏä§Ìä∏ Îç∞Ïù¥ÌÑ∞  ùëãtest Ïùò ÌÅ¥ÎûòÏä§Î•º ÏòàÏ∏°Ìï† ÎïåÎäî Í∞ÄÏû• ÏÉÅÏúÑÏùò ÎÖ∏ÎìúÎ∂ÄÌÑ∞ Î∂ÑÎ•òÍ∑úÏπôÏùÑ Ï∞®Î°ÄÎåÄÎ°ú Ï†ÅÏö©ÌïòÏó¨ ÎßàÏßÄÎßâÏóê ÎèÑÎã¨ÌïòÎäî ÎÖ∏ÎìúÏùò Ï°∞Í±¥Î∂Ä ÌôïÎ•† Î∂ÑÌè¨Î•º Ïù¥Ïö©ÌïòÏó¨ ÌÅ¥ÎûòÏä§Î•º ÏòàÏ∏°ÌïúÎã§.

$$\hat{Y} = \text{arg}\max_k P(Y=k|X_{\text{test}})_{\text{last node}}$$


## Î∂ÑÎ•òÍ∑úÏπôÏùÑ Ï†ïÌïòÎäî Î∞©Î≤ï

#### Î∂ÑÎ•ò Í∑úÏπôÏùÑ Ï†ïÌïòÎäî Î∞©Î≤ïÏùÄ Î∂ÄÎ™® ÎÖ∏ÎìúÏôÄ ÏûêÏãù ÎÖ∏Îìú Í∞ÑÏùò ÏóîÌä∏Î°úÌîºÎ•º Í∞ÄÏû• ÎÇÆÍ≤å ÎßåÎìúÎäî ÏµúÏÉÅÏùò ÎèÖÎ¶Ω Î≥ÄÏàòÏôÄ Í∏∞Ï§ÄÍ∞íÏùÑ Ï∞æÎäî Í≤ÉÏù¥Îã§. 

Ïù¥Îü¨Ìïú Í∏∞Ï§ÄÏùÑ Ï†ïÎüâÌôîÌïú Í≤ÉÏù¥ Ï†ïÎ≥¥ÌöçÎìùÎüâ(information gain)Ïù¥Îã§. Í∏∞Î≥∏Ï†ÅÏúºÎ°ú Î™®Îì† ÎèÖÎ¶Ω Î≥ÄÏàòÏôÄ Î™®Îì† Í∞ÄÎä•Ìïú Í∏∞Ï§ÄÍ∞íÏóê ÎåÄÌï¥ Ï†ïÎ≥¥ÌöçÎìùÎüâÏùÑ Íµ¨ÌïòÏó¨ Í∞ÄÏû• Ï†ïÎ≥¥ÌöçÎìùÎüâÏù¥ ÌÅ∞ ÎèÖÎ¶Ω Î≥ÄÏàòÏôÄ Í∏∞Ï§ÄÍ∞íÏùÑ ÏÑ†ÌÉùÌïúÎã§.

## Ï†ïÎ≥¥ÌöçÎìùÎüâ

#### Ï†ïÎ≥¥ÌöçÎìùÎüâ(information gain)Îäî  ùëã ÎùºÎäî Ï°∞Í±¥Ïóê ÏùòÌï¥ ÌôïÎ•† Î≥ÄÏàò  ùëå Ïùò ÏóîÌä∏Î°úÌîºÍ∞Ä ÏñºÎßàÎÇò Í∞êÏÜåÌïòÏòÄÎäîÍ∞ÄÎ•º ÎÇòÌÉÄÎÇ¥Îäî Í∞íÏù¥Îã§. 

Îã§ÏùåÍ≥º Í∞ôÏù¥  ùëå Ïùò ÏóîÌä∏Î°úÌîºÏóêÏÑú  ùëã Ïóê ÎåÄÌïú  ùëå Ïùò Ï°∞Í±¥Î∂Ä ÏóîÌä∏Î°úÌîºÎ•º Î∫Ä Í∞íÏúºÎ°ú Ï†ïÏùòÎêúÎã§.

$$IG[Y,X] = H[Y] - H[Y|X]$$

ÏòàÎ•º Îì§Ïñ¥ A, B Îëê Í∞ÄÏßÄÏùò Îã§Î•∏ Î∂ÑÎ•ò Í∑úÏπôÏùÑ Ï†ÅÏö©ÌñàÎçîÎãà Îã§ÏùåÍ≥º Í∞ôÏù¥ ÏÑúÎ°ú Îã§Î•¥Í≤å Îç∞Ïù¥ÌÑ∞Í∞Ä ÎÇòÎâòÏñ¥ Ï°åÎã§Í≥† Í∞ÄÏ†ïÌïòÏûê.

[Í∑∏Î¶º] Î∂ÑÎ•òÍ≤∞Í≥º ÏòàÏãú: https://datascienceschool.net/upfiles/b9fe188ad28b4c9fb2cd5a0292b49502.png

A Î∞©Î≤ïÍ≥º B Î∞©Î≤ï Î™®Îëê ÎÖ∏Îìú Î∂ÑÎ¶¨ Ï†ÑÏóêÎäî Y=0 Ïù∏ Îç∞Ïù¥ÌÑ∞Ïùò ÏàòÏôÄ Y=1 Ïù∏ Îç∞Ïù¥ÌÑ∞Ïùò ÏàòÍ∞Ä Î™®Îëê 40Í∞úÏòÄÎã§.

A Î∞©Î≤ïÏúºÎ°ú ÎÖ∏ÎìúÎ•º Î∂ÑÎ¶¨ÌïòÎ©¥ Îã§ÏùåÍ≥º Í∞ôÏùÄ Îëê Í∞úÏùò ÏûêÏãù ÎÖ∏ÎìúÍ∞Ä ÏÉùÍ∏¥Îã§.

+ ÏûêÏãù ÎÖ∏Îìú A1ÏùÄ Y=0 Ïù∏ Îç∞Ïù¥ÌÑ∞Í∞Ä 30Í∞ú, Y=1 Ïù∏ Îç∞Ïù¥ÌÑ∞Í∞Ä 10Í∞ú
+ ÏûêÏãù ÎÖ∏Îìú A2ÏùÄ Y=0 Ïù∏ Îç∞Ïù¥ÌÑ∞Í∞Ä 10Í∞ú, Y=1 Ïù∏ Îç∞Ïù¥ÌÑ∞Í∞Ä 30Í∞ú

B Î∞©Î≤ïÏúºÎ°ú ÎÖ∏ÎìúÎ•º Î∂ÑÎ¶¨ÌïòÎ©¥ Îã§ÏùåÍ≥º Í∞ôÏùÄ Îëê Í∞úÏùò ÏûêÏãù ÎÖ∏ÎìúÍ∞Ä ÏÉùÍ∏¥Îã§.

+ ÏûêÏãù ÎÖ∏Îìú B1ÏùÄ Y=0 Ïù∏ Îç∞Ïù¥ÌÑ∞Í∞Ä 20Í∞ú, Y=1 Ïù∏ Îç∞Ïù¥ÌÑ∞Í∞Ä 40Í∞ú
+ ÏûêÏãù ÎÖ∏Îìú B2ÏùÄ Y=0 Ïù∏ Îç∞Ïù¥ÌÑ∞Í∞Ä 20Í∞ú, Y=1 Ïù∏ Îç∞Ïù¥ÌÑ∞Í∞Ä 0Í∞ú

Ïö∞ÏÑ† Î∂ÄÎ™® ÎÖ∏ÎìúÏùò ÏóîÌä∏Î°úÌîºÎ•º Í≥ÑÏÇ∞ÌïòÎ©¥ Îã§ÏùåÍ≥º Í∞ôÎã§.

$$H[Y] = -\dfrac{1}{2}\log_2\left(\dfrac{1}{2}\right) -\dfrac{1}{2}\log_2\left(\dfrac{1}{2}\right) = \dfrac{1}{2} + \dfrac{1}{2}  = 1$$

A Î∞©Î≤ïÏóê ÎåÄÌï¥ IGÎ•º Í≥ÑÏÇ∞ÌïòÎ©¥ Îã§ÏùåÍ≥º Í∞ôÎã§.

$$H[Y|X=X_1] = -\dfrac{3}{4}\log_2\left(\dfrac{3}{4}\right) -\dfrac{1}{4}\log_2\left(\dfrac{1}{4}\right) = 0.81$$

$$H[Y|X=X_2] = -\dfrac{1}{4}\log_2\left(\dfrac{1}{4}\right)  -\dfrac{3}{4}\log_2\left(\dfrac{3}{4}\right) = 0.81$$

$$H[Y|X] = \dfrac{1}{2} H[Y|X=X_1] + \dfrac{1}{2} H[Y|X=X_2] = 0.81$$

$$IG = H[Y] - H[Y|X] = 0.19$$

B Î∞©Î≤ïÏóê ÎåÄÌï¥ IGÎ•º Í≥ÑÏÇ∞ÌïòÎ©¥ Îã§ÏùåÍ≥º Í∞ôÎã§.

$$H[Y|X=X_1] = -\dfrac{1}{3}\log_2\left(\dfrac{1}{3}\right) - \dfrac{2}{3}\log_2\left(\dfrac{2}{3}\right) = 0.92$$

$$H[Y|X=X_2] = 0$$

$$H[Y|X] = \dfrac{3}{4} H[Y|X=X_1] + \dfrac{1}{4} H[Y|X=X_2] = 0.69$$

$$IG = H[D] - H[Y|X] = 0.31$$

Îî∞ÎùºÏÑú B Î∞©Î≤ïÏù¥ Îçî ÎÇòÏùÄ Î∞©Î≤ïÏûÑÏùÑ Ïïå Ïàò ÏûàÎã§.

## Scikit-LearnÏùò ÏùòÏÇ¨Í≤∞Ï†ïÎÇòÎ¨¥ ÌÅ¥ÎûòÏä§

Scikit-LearnÏóêÏÑú ÏùòÏÇ¨Í≤∞Ï†ïÎÇòÎ¨¥Îäî DecisionTreeClassifier ÌÅ¥ÎûòÏä§Î°ú Íµ¨ÌòÑÎêòÏñ¥ ÏûàÎã§. Ïó¨Í∏∞ÏóêÏÑúÎäî Î∂ìÍΩÉ Î∂ÑÎ•òÎ¨∏Ï†úÎ•º ÏòàÎ•º Îì§Ïñ¥ ÏùòÏÇ¨Í≤∞Ï†ïÎÇòÎ¨¥Î•º ÏÑ§Î™ÖÌïúÎã§. Ïù¥ ÏòàÏ†úÏóêÏÑúÎäî ÎèÖÎ¶ΩÎ≥ÄÏàò Í≥µÍ∞ÑÏùÑ Í≥µÍ∞ÑÏÉÅÏóê ÌëúÏãúÌïòÍ∏∞ ÏúÑÌï¥ ÍΩÉÏùò Í∏∏Ïù¥ÏôÄ Ìè≠ÎßåÏùÑ ÎèÖÎ¶ΩÎ≥ÄÏàòÎ°ú ÏÇ¨Ïö©ÌïòÏòÄÎã§.

```python
from sklearn.datasets import load_iris

iris = load_iris()
y = iris.target
X = iris.data[:, 2:] # ÍΩÉÏïû Í∏∏Ïù¥ÏôÄ Ìè≠ ÎèÖÎ¶ΩÎ≥ÄÏàòÎßå ÏÇ¨Ïö©
feature_names = iris.feature_names[2:]

from sklearn.tree import DecisionTreeClassifier

tree1 = DecisionTreeClassifier(criterion='entropy', max_depth=1, random_state=0).fit(X, y)
```

Îã§ÏùåÏùÄ ÏùòÏÇ¨Í≤∞Ï†ïÎÇòÎ¨¥Î•º ÏãúÍ∞ÅÌôîÌïòÍ∏∞ ÏúÑÌïú ÏΩîÎìúÏù¥Îã§. 

draw_decision_tree Ìï®ÏàòÎäî ÏùòÏÇ¨Í≤∞Ï†ïÎÇòÎ¨¥Ïùò ÏùòÏÇ¨ Í≤∞Ï†ï Í≥ºÏ†ïÏùò ÏÑ∏Î∂ÄÏ†ÅÏù∏ ÎÇ¥Ïó≠ÏùÑ Îã§Ïù¥Ïñ¥Í∑∏Îû®ÏúºÎ°ú Î≥¥Ïó¨Ï£ºÍ≥† plot_decision_regions Ìï®ÏàòÎäî Ïù¥Îü¨Ìïú ÏùòÏÇ¨ Í≤∞Ï†ïÏóê ÏùòÌï¥ Îç∞Ïù¥ÌÑ∞Ïùò ÏòÅÏó≠Ïù¥ Ïñ¥ÎñªÍ≤å ÎÇòÎâòÏñ¥Ï°åÎäîÏßÄÎ•º ÏãúÍ∞ÅÌôîÌïòÏó¨ Î≥¥Ïó¨Ï§ÄÎã§.

```python
import io
import pydot
from IPython.core.display import Image
from sklearn.tree import export_graphviz
```

```python
import matplotlib.pylab as plt
import matplotlib as mpl
import matplotlib.font_manager as fm
import seaborn as sns
import numpy as np

sns.set_style("whitegrid")
mpl.rcParams['axes.unicode_minus'] = False
plt.rcParams['font.size'] = 15

path = "/Library/Fonts/NanumGothic.otf"
font_name = fm.FontProperties(fname=path, size=20).get_name()

plt.rc('font', family=font_name)
```

```python
def draw_decision_tree(model):
    dot_buf = io.StringIO()
    export_graphviz(model, out_file=dot_buf, feature_names=feature_names)
    graph = pydot.graph_from_dot_data(dot_buf.getvalue())[0]
    image = graph.create_png()
    return Image(image)

def plot_decision_regions(X, y, model, title):
    resolution = 0.01
    markers = ('s', '^', 'o')
    colors = ('red', 'blue', 'lightgreen')
    cmap = mpl.colors.ListedColormap(colors)
    
    x1_min, x1_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    x2_min, x2_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx1, xx2 = np.meshgrid(np.arange(x1_min, x1_max, resolution), np.arange(x2_min, x2_max, resolution))
    Z = model.predict(np.array([xx1.ravel(), xx2.ravel()]).T).reshape(xx1.shape)
    
    plt.figure(figsize=(8, 5))
    plt.contour(xx1, xx2, Z, cmap=mpl.colors.ListedColormap(['k']))
    plt.contourf(xx1, xx2, Z, alpha=0.3, cmap=cmap)
    plt.xlim(xx1.min(), xx1.max())
    plt.ylim(xx2.min(), xx2.max())
    
    for idx, cl in enumerate(np.unique(y)):
        plt.scatter(x=X[y==cl, 0], y=X[y==cl, 1], alpha=0.8, c=[cmap(idx)], s=80, label=cl, marker=markers[idx])
        
    plt.xlabel(iris.feature_names[2])
    plt.ylabel(iris.feature_names[3])
    plt.legend(loc="upper left")
    plt.title(title, y=1.1)
    
    return Z
```

```python
draw_decision_tree(tree1)
```

```python
plot_decision_regions(X, y, tree1, "Depth 1")
plt.show()
```

```python
from sklearn.metrics import confusion_matrix

confusion_matrix(y, tree1.predict(X))
```

(2) K=5 Ïù∏ ÍµêÏ∞® Í≤ÄÏ¶ùÏùÑ ÌÜµÌï¥ ÌÖåÏä§Ìä∏ ÏÑ±Îä• ÌèâÍ∑†ÏùÑ Ï∏°Ï†ïÌïòÎùº.

(3) max_depth Ïù∏ÏàòÎ•º Î∞îÍæ∏Ïñ¥ Í∞ÄÎ©¥ÏÑú ÌÖåÏä§Ìä∏ ÏÑ±Îä• ÌèâÍ∑†ÏùÑ Íµ¨ÌïòÏó¨ cross validation curveÎ•º Í∑∏Î¶¨Í≥† Í∞ÄÏû• ÌÖåÏä§Ìä∏ ÏÑ±Îä• ÌèâÍ∑†Ïù¥ Ï¢ãÏùÄ max_depth Ïù∏ÏàòÎ•º Ï∞æÏïÑÎùº.

```python
X = iris.data[:, :2]
y = iris.target
```

```python
tree1 = DecisionTreeClassifier(criterion="entropy", max_depth=3, random_state=0).fit(X, y)
```

```python
draw_decision_tree(tree1)
```

```python
plot_decision_regions(X, y, tree1, "Depth 3")
plt.show()
```

```python
confusion_matrix(y, tree1.predict(X))
```

```python
# Ï†ïÌôïÎèÑ 0.8
from sklearn.metrics import classification_report
print(classification_report(y, tree1.predict(X), target_names=iris.target_names))
```

```python
# K = 5  ÍµêÏ∞®Í≤ÄÏ¶ù ÌÖåÏä§Ìä∏

from sklearn.model_selection import KFold, cross_val_score, validation_curve

tree1 = DecisionTreeClassifier(criterion="entropy", max_depth=3, random_state=0).fit(X, y)
cv = KFold(5, shuffle=True, random_state=0)
score = cross_val_score(tree1, X, y, scoring="accuracy", cv=cv)
```

```python
# ÏÑ±Îä•ÌèâÍ∑†
score.mean()
```

```python
for depth in range(10):
    model = DecisionTreeClassifier(criterion="entropy", max_depth=(depth+1), random_state=0).fit(X, y)
    cv = KFold(5, shuffle=True, random_state=0)
    score = cross_val_score(model, X, y, scoring="accuracy", cv=cv)
    print("Depth{} : ÏÑ±Îä•Ï†êÏàò {}".format((depth+1), score.mean()))
```

Í∞ÄÏû• ÏÑ±Îä•Ïù¥ Ï¢ãÏùÄ max_depthÎäî 5Ïù¥Îã§.

```python
param_range = np.arange(1, 11)
train_scores, test_scores = \
    validation_curve(DecisionTreeClassifier(), X, y, param_name='max_depth', param_range=param_range, scoring="accuracy", cv=5)
```

```python
train_scores_mean = np.mean(train_scores, axis=1)
train_scores_std = np.std(train_scores, axis=1)
test_scores_mean = np.mean(test_scores, axis=1)
test_scores_std = np.std(test_scores, axis=1)

param_range = np.arange(1, 11)
plt.title("Validation Curve with Decision Tree")
plt.xlabel("max_depth")
plt.ylabel("Score")
plt.ylim(0.1, 1.1)
plt.semilogx(param_range, train_scores_mean, label="Training Score", color="darkorange", lw=2)
plt.fill_between(param_range, train_scores_mean-train_scores_std, train_scores_mean+train_scores_std, alpha=0.2, color="darkorange", lw=2)
plt.semilogx(param_range, test_scores_mean, label="Cross-validation score", color="navy", lw=2)
plt.fill_between(param_range, test_scores_mean-test_scores_std, test_scores_mean+test_scores_std, alpha=0.2, color="navy", lw=2)
plt.legend(loc="best")
plt.xticks(param_range)
plt.show()
```

## ÌöåÍ∑Ä ÎÇòÎ¨¥

ÏòàÏ∏°Í∞í  ùë¶ÃÇ  ÏùÑ Îã§ÏùåÍ≥º Í∞ôÏù¥ Í∞Å ÌäπÏßïÍ∞í ÏòÅÏó≠ÎßàÎã§ Í≥†Ï†ïÎêú Í∞í  ùë¶1, ùë¶2 Î•º ÏÇ¨Ïö©ÌïòÍ≥†,

$$\hat{y} = 
\begin{cases} 
y_1 & \text{ if } x \geq x_{\text{threshold}} \\ 
y_2 & \text{ if } x < x_{\text{threshold}}
\end{cases}$$

#### Í∏∞Ï§ÄÍ∞í Î∞è  ùë¶1,ùë¶2 Î•º ÏÑ†ÌÉùÌïòÎäî Î™©Ï†ÅÌï®ÏàòÎ°ú Ïò§Ï∞® Ï†úÍ≥±Ìï©ÏùÑ ÏÇ¨Ïö©ÌïòÎ©¥ ÌöåÍ∑ÄÎ∂ÑÏÑùÎ•º Ìï† Ïàò ÏûàÎã§. Ïù¥Îü¨Ìïú Î™®ÌòïÏùÑ ÌöåÍ∑Ä ÎÇòÎ¨¥(regression tree)ÎùºÍ≥† ÌïúÎã§.

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
rng = np.random.RandomState(1)  # RandomStateÎäî Îã§ÏñëÌïú ÌôïÎ•†Î∂ÑÌè¨Ï∏°Î©¥Ïóê Ïàò ÎßéÏùÄ ÎûúÎç§ÎÑòÎ≤Ñ ÏÉùÏÑ±Í∏∞Îì§ÏùÑ Í∞ÄÏßÄÍ≥† ÏûàÎã§.
X = np.sort(5 * rng.rand(80, 1), axis=0)  # 0-1ÏÇ¨Ïù¥Ïùò ÎÇúÏàò 80 x 1 ÏÉùÏÑ±
y = np.sin(X).ravel()
y[::5] += 3 * (0.5 - rng.rand(16)) # 5Í∞ú Í∞ÑÍ≤©ÏúºÎ°ú Ïà´Ïûê Î≥ÄÍ≤Ω

regtree = DecisionTreeRegressor(max_depth=3)
regtree.fit(X, y)

X_test = np.arange(0.0, 5.0, 0.01)[:, np.newaxis]
y_hat = regtree.predict(X_test)
```

[Ï∞∏Í≥†] rng = numpy.random.RandomState(123)Í≥º Í∞ôÏù¥ ÌÅ¥ÎûòÏä§ Î≥ÄÏàò rngÎ•º ÏÑ†Ïñ∏ Ìï† Ïàò ÏûàÎã§. Ïù¥Îïå 123ÏùÄ ÏùºÏ¢ÖÏùò Î≤àÌò∏Î°ú randomstate Î≥ÄÏàòÎ•º ÏÉùÏÑ±Ìï† Îïå Î≤àÌò∏Î•º ÏßÄÏ†ïÌï¥ÏÑú ÎÑ£Ïñ¥Ï£ºÏßÄ ÏïäÏúºÎ©¥, Ïó¨Îü¨Í∞úÏùò randomstate Î≥ÄÏàòÎ•º ÏÇ¨Ïö©Ìï† Îïå Î™®Îëê Í∞ôÏùÄ randomÌïú Í∞íÏùÑ Ï∂úÎ†•ÌïòÍ≤å ÎêúÎã§. Îî∞ÎùºÏÑú ÏÑ†Ïñ∏ÏãúÏóê Í∞Å Î≥ÄÏàòÎßàÎã§ Îã§Î•∏ Í∞íÏùÑ ÎÑ£ÏùåÏúºÎ°úÏç® ÌïòÎÇòÏùò ÏΩîÎìúÏÉÅÏóêÏÑú Îã§Î•∏ ÎûúÎç§Ìïú Í∞íÏùÑ Ï∂úÎ†•Ìï† Ïàò ÏûàÍ≤å ÎêúÎã§.

```python
# Plot the results
plt.figure()
plt.scatter(X, y, s=20, edgecolors="black", c="darkorange", label="Îç∞Ïù¥ÌÑ∞")
plt.plot(X_test, y_hat, color="cornflowerblue", lw=2, label="ÏòàÏ∏°")
plt.xlabel("x")
plt.ylabel("$y$ & $\hat{y}$")
plt.title("ÌöåÍ∑ÄÎÇòÎ¨¥")
plt.legend()
plt.show()
```
